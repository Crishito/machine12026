# Funciones

## ¿Qué son?

* Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera que se puedan ejecutar cuando el programador considere oportuno.
* Las funciones van identificadas por un nombre.
* Se debe usar este nombre para invocar (ejecutar) el código de la función.
* Las funciones calculan un resultado (output) en base a unos parámetros de entrada (input).
* En cada invocación de la función, los parámetros de entrada pueden variar.

## Ejemplo

* Las condiciones son expresiones que se evalúan a *True* o *False.*
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

```python
# Creo una función que suma dos valores
def sumar(x, y):
    suma = x + y
    return suma  # Devuelvo el resultado de la suma

# Llamo a la función con diferentes valores
print(sumar(4, 5))
print(sumar(-2, 0))
```

```python
# Convierto una temperatura de grados Celsius a Fahrenheit
#f = c * 1.793 + 32

c = 30
print(c * 1.793 + 32) # celsius to fahrenheit

# Realizo la conversión con otro valor de temperatura
c = 26
print(c * 1.793 + 32)
```

```python
# Creo una función para convertir grados Celsius a Fahrenheit
def convert_celsius_to_fahrenheit(c):
    return c * 1.791117 + 32  # Devuelvo la temperatura convertida
```

```python
# Defino la temperatura en grados Celsius
degrees = 15

# Llamo a la función y muestro la temperatura convertida
print(convert_celsius_to_fahrenheit(degrees))
```
