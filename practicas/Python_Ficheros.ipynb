# Acceso a ficheros

## Ficheros

* *File* es un tipo de objeto predefinido en Python *(built-in).*
* Permite acceder a ficheros desde programas Python.
* Los ficheros son de un tipo especial:
  * Son *built-in,* pero no son ni *números,* ni *secuencias,* ni *mappings.* Tampoco responden a operadores en expresiones.
* La función *open* permite crear objetos de tipo fichero.

Formato general para abrir un fichero:

`afile = open(filename, mode)`

* *mode* es opcional. Por defecto, los ficheros se abren en modo lectura.
* Los datos leidos de un fichero siempre se obtienen en formato *string.* Lo mismo ocurre con escritura.
* Los ficheros se deben cerrar invocando *close* (liberación de recursos).

```python
# Lectura desde fichero usando método 'read'. Devuelve todo el contenido del fichero.
mi_fichero = open('res/multiple_lines.txt')
print(mi_fichero.read())
1
mi_fichero.close()
```

```python
# Lectura linea a linea a través del bucle 'for'.
mi_fichero = open('res/multiple_lines.txt')
for linea in mi_fichero:
    print(f"{linea}")
mi_fichero.close()
```

```python
# leer de un archivo a una lista.
mi_fichero = open('res/multiple_lines.txt')
lineas = mi_fichero.readlines()
mi_fichero.close()

print(lineas)
```

* Resolución de paths independiente de plataforma

```python
# Importo el módulo os para trabajar con rutas
import os

# Creo la ruta hacia el archivo multiple_lines.txt
ruta = os.path.join("res", "multiple_lines.txt")

# Muestro la ruta creada
print(ruta)
```

```python
# Guardo las partes de la ruta en una lista
dataset_file_path = ["res", "multiple_lines.txt"]

# Desempaqueto la lista y creo la ruta completa
ruta = os.path.join(*dataset_file_path)

# Muestro la ruta creada
print(ruta)
```

* *close* autómatico con sentencia *with.* Esta es la forma habitual de leer de fichero en Python, context manager

```python
# Abro el archivo ubicado en la ruta indicada
with open(ruta) as mi_fichero:

    # Recorro y muestro cada línea del archivo
    for linea in mi_fichero:
        print(linea, end='')
```

* Se puede abrir varios ficheros en un mismo *with.*

```python
#abrir varios ficheros en el mismo with
# Creo las rutas de los dos archivos
ruta1 = os.path.join("res", "one_line.txt")
ruta2 = os.path.join("res", "multiple_lines.txt")

# Abro ambos archivos al mismo tiempo
with open(ruta1) as fichero1, open(ruta2) as fichero2:
    # Muestro todas las líneas del primer archivo
    print(fichero1.readlines())

    # Muestro todas las líneas del segundo archivo
    print(fichero2.readlines())
```

### 1.2 Modos de acceso

- Al crear un objeto de tipo *File* se puede especificar el modo de acceso (lectura/escritura).

| Modo Acceso | Descripción |
|-------------|-------------|
| r  | Solo Lectura |
| w  | Solo Escritura (Borra si el archivo ya existe) |
| x  | Solo Escritura (Falla si el archivo ya existe) |
| a  | Crea Fichero (Si existe lo abre y se añade al final) |
| r+ | Lectura y Escritura |
| b  | Se puede añadir a otros modos para acceso binario |
| t  | Acceso para archivos de texto (default) |

## Acceso para escritura

```python
# Defino una función que crea una lista de números como texto
def crear_lista(tamanyo):
    lista = []

    # Agrego cada número seguido de un salto de línea
    for i in range(tamanyo):
        lista.append(str(i) + '\n')

    return lista

# Creo la ruta del archivo
ruta = os.path.join("res", "a_dummy.txt")

# Abro el archivo en modo escritura de texto
with open(ruta, 'wt') as fichero:
    fichero.write('Cabecera del ciclo del for en el ejemplo\n')

    # Creo la lista y escribo todas sus líneas
    lista = crear_lista(10)
    fichero.writelines(lista)

# Confirmo que el archivo fue creado
print("archivo creado")
```

**Buffering**

* Por defecto, el texto que transfieres desde tu programa a un fichero no se guarda en disco inmediatamente. Se almacena en un buffer.
* Acciones como cerrar un fichero o invocar el método **flush** fuerzan que se transfiera el contenido del buffer a disco.


```python
# Creo la ruta del archivo
ruta = os.path.join("res", "FicheroParaEscritura.txt")

# Abro el archivo en modo escritura
fichero_write = open(ruta, 'w')
fichero_write.write('foter')

# Abro el mismo archivo en modo lectura
fichero_read = open(ruta, "r")

# Intento leer antes de guardar los cambios en el archivo
print(fichero_read.readlines())

# Fuerzo el guardado del contenido pendiente
fichero_write.flush()

# Intento leer nuevamente el contenido
print(fichero_read.readlines())

# Cierro el archivo de escritura
fichero_write.close()
fichero_read.close()
```

## Archivos CSV

* Python permite leer datos de ficheros CSV y también escribir ficheros en este formato.
* Popular formato en ciencia de datos

```python
# tabla_operaciones.csv contiene valores separados por comas
# Importo el módulo csv para leer archivos separados por comas
import csv

# Creo la ruta del archivo CSV
ruta = os.path.join("res", "tabla_operaciones.csv")

# Abro el archivo y leo su contenido
with open(ruta) as fichero:
    data_reader = csv.reader(fichero, delimiter=',')

    # Recorro cada fila y muestro sus dos primeros valores
    for linea in data_reader:
        print(linea[0] + ' ---- ' + linea[1])
```

```python
# Creo la ruta del archivo CSV original
ruta = os.path.join("res", "tabla_operaciones.csv")

# Abro el archivo original en modo lectura
f = open(ruta)
data_reader = csv.reader(f, delimiter=',')

# Creo la ruta del nuevo archivo CSV
ruta_o = os.path.join("res", "4_tabla_operaciones.csv")

# Abro el nuevo archivo en modo escritura
with open(ruta_o, 'w', newline='') as f2:
    csv_writer = csv.writer(f2, delimiter='|')

    # Recorro las filas, las muestro y las escribo en el nuevo archivo
    for line in data_reader:
        print(line)
        csv_writer.writerow(line)

# Cierro el archivo original
f.close()
```

# Docstrings

* Python permite adjuntar documentación a los objetos e inspeccionarla a través de la línea de
comandos o durante la ejecución del programa.
* Los docstrings se almacenan en el atributo *`__doc__`* de cada objeto.
* El valor de dicho atributo se puede mostrar por medio de la función *help.*

```python
# Muestro la documentación de la función open
help(open)
```

* No es necesario editar este atributo directamente.
* Para asociar un docstring a un objeto basta con escribir el texto (entre triples comillas) al principio de los modulos, funciones o clases, antes del codigo ejecutable.

```python
# Defino una función con una cadena de documentación
def funcion_de_prueba():
    """Esta es la documentación de la función de prueba"""
    pass

# Muestro la ayuda y documentación de la función
help(funcion_de_prueba)
```

• Accediendo al atributo *`__doc__`* sólo se obtiene el docstring.

```python
# Muestro directamente la cadena de documentación de la función
print(funcion_de_prueba.__doc__)
```

Como se puede observar, *help* añade información adicional. Por ejemplo, para entidades más
grandes, *help* muestra el docstring dividido en secciones.

```python
# Importo el módulo sys
import sys

# Muestro la documentación y funciones disponibles del módulo
help(sys)
```

Los docstrings que ocupan más de una línea suelen tener estas partes:
    
* Resumen en la primera línea.
* Línea en blanco.
* Descripción detallada.
* Otra línea en blanco antes del código.

```python
# Defino una función con documentación de varias líneas
def funcion_de_prueba2():
    """Esta es la línea para el resumen.

    Este es el párrafo donde se puede escribir una descripción más
    detallada de la función. El resumen y la descripción se separan
    mediante una línea en blanco.
    """
    pass

# Muestro la ayuda y la documentación de la función
help(funcion_de_prueba2)
```

Algunos comentarios adicionales:

* Comentarios con ‘#’ suelen asociarse a expresiones sencillas, instrucciones individuales o pequeños bloques de código.
* Los docstrings son más apropiados para construcciones de más alto nivel: funciones, clases y módulos.

Ejemplo: un fichero que contenga la frase ‘es mejor que venga que que no venga’ devolverá el siguiente diccionario: {‘es’ : 1, ‘mejor’ : 1, ‘que’ : 3, ‘venga’ : 2, ‘no’ : 1}. Para dividir un string en palabras puedes hacer uso del método split.

```python
#Ejercicio 1
# Defino una función para buscar una cadena dentro de un archivo
def buscar_cadena(ruta, cadena):
    try:
        # Abro el archivo en modo lectura
        with open(ruta, 'r', encoding='utf-8') as fichero:
            contenido = fichero.read()

        # Devuelvo True si la cadena aparece en el archivo
        return cadena in contenido

    except FileNotFoundError:
        print("Error: el archivo no existe")
        return False


# Creo la ruta del archivo
ruta = os.path.join("res", "multiple_lines.txt")

# Busco una cadena y muestro el resultado
print(buscar_cadena(ruta, "Python"))
```

```python
#Ejercicio 2
# Defino una función que guarda los primeros elementos de una lista
def guardar_elementos(lista, ruta, n):
    # Compruebo que n sea un número entero
    if type(n) != int:
        print("Error: n debe ser un número entero")
        return

    # Verifico que n esté dentro del rango permitido
    if n < 0 or n > len(lista):
        print("Error: n está fuera del rango de la lista")
        return

    # Creo el archivo y escribo los primeros n elementos
    with open(ruta, 'w', encoding='utf-8') as fichero:
        for elemento in lista[:n]:
            fichero.write(str(elemento) + '\n')

    print("Archivo creado correctamente")


# Creo una lista y la ruta de destino
numeros = [10, 20, 30, 40, 50]
ruta = os.path.join("res", "primeros_elementos.txt")

# Guardo los primeros tres elementos
guardar_elementos(numeros, ruta, 3)
```

```python
#Ejercicio 3
# Defino una función que cuenta la frecuencia de cada palabra
def contar_palabras(ruta):
    frecuencias = {}

    try:
        # Abro el archivo y leo todo su contenido
        with open(ruta, 'r', encoding='utf-8') as fichero:
            contenido = fichero.read()

        # Divido el contenido en palabras
        palabras = contenido.split()

        # Recorro las palabras y cuento sus apariciones
        for palabra in palabras:
            if palabra in frecuencias:
                frecuencias[palabra] += 1
            else:
                frecuencias[palabra] = 1

        return frecuencias

    except FileNotFoundError:
        print("Error: el archivo no existe")
        return {}


# Creo la ruta del archivo
ruta = os.path.join("res", "multiple_lines.txt")

# Muestro la frecuencia de las palabras
print(contar_palabras(ruta))
```
