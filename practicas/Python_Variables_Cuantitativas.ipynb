```python
# Creación aleatoria de datos
# Importo NumPy
import numpy as np

# Creo un arreglo de datos
X = np.array([0.5, 23, 0.3, 4.5, 0.3, 0.5, -28, -50, 60, -100, -10, -11, 13, 19, 1, 9])

print(X)
```

```python
# Medidas de tendencia central
# Importo la librería
from scipy import stats

# Calculo las medidas de tendencia central
media = np.mean(X)
mediana = np.median(X)
moda, _ = stats.mode(X)

print('media: ', media)
print('mediana: ', mediana)
print('moda :', moda)
```

```python
# Medidas de posición
# Importo las librerías
import numpy as np
import matplotlib.pyplot as plt

# Calculo las medidas de posición
Q1 = np.percentile(X, 25)
Q3 = np.percentile(X, 75)
RIC = Q3 - Q1
lim_inf = Q1 - 1.5 * RIC
lim_sup = Q3 + 1.5 * RIC

print('RIC :', RIC)
print('limite inferior :', lim_inf)
print('limite superior :', lim_sup)
```

```python
# Dibuja el diagrama de box and whiskers y determina visualmente el número de outliers
# Creo el diagrama de caja y bigotes
plt.boxplot(X)

plt.grid()
plt.title("Diagrama de caja y bigotes")
plt.show()
```

```python
# Medidas de dispersión
# Calculo las medidas de dispersión
desvest = np.std(X)

print('Desviación típica: ', desvest)
print('Varianza: ', desvest**2)
```

```python
# Medidas de distribución
# Calculo las medidas de distribución
asimetria = stats.skew(X)
curtosis = stats.kurtosis(X, fisher=True) # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.htm

print('Asimetría ', np.round(asimetria, 4))
print('Curtosis ', np.round(curtosis, 4))

# Verifico el tipo de distribución
if curtosis > 0:
    print('Distribución Leptocúrtica')
elif curtosis < 0:
    print('Distribución Platicúrtica')
else:
    print('Distribución Mesocúrtica')
```

# EJERCICIO EXTRA SOBRE EL DATASET DE IRIS
# Extraer la media y la desviación típica de cada variable en función de la clase y representarlo en un gráfico (plt.errorbar)

```python
# Importo las librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# Cargo el conjunto de datos Iris
iris = datasets.load_iris()

# Asigno los datos y las clases
X = iris.data
y = iris.target

# Recorro cada clase
for clase in np.unique(y):
    datos = X[y == clase]

    # Calculo la media y la desviación
    media = np.mean(datos, axis=0)
    desvest = np.std(datos, axis=0)

    # Grafico la media con barras de error
    plt.errorbar(
        iris.feature_names,
        media,
        yerr=desvest,
        marker='o',
        capsize=5,
        label=iris.target_names[clase]
    )

# Asigno un título y etiquetas
plt.title("Media y desviación estándar de cada variable por especie (Iris)")
plt.xlabel("Variables")
plt.ylabel("Valor (cm)")

# Giro las etiquetas del eje X
plt.xticks(rotation=45)

plt.legend()
plt.grid(True)

plt.show()
```
