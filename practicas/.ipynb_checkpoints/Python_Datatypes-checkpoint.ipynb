## 2. Literales, variables y tipos de datos básicos

De forma muy genérica, al ejecutarse un programa Python, simplemente se realizan operaciones sobre objetos. Estos dos términos son fundamentales:
* **Objetos:** Cualquier tipo de datos (números, caracteres o datos más complejos).
* **Operaciones:** Cómo manipulamos estos datos.

```python
4 + 3
```

### 2.1 Literales
Python tiene una serie de tipos de datos integrados en el propio lenguaje. Los literales son expresiones que generan objetos de estos tipos. Estos objetos, según su tipo, pueden ser:
* Simples o compuestos.
* Mutables o inmutables.

#### Literales simples
* Enteros
* Decimales o punto flotante
* Booleano

```python
print(4)          # número entero
print(4.2)        # número en coma flotante
print('Hello world!') # string
print(False)      # boolean
```

#### Literales compuestos
* **Listas:** Mutables `[1, 2, 3]`
* **Diccionarios:** Mutables `{'Nombre': 'John Doe', 'edad': 30}`
* **Conjuntos:** Mutables `{1, 2, 3}`
* **Tuplas:** Inmutables `(4, 5)`

```python
print([1, 2, 3])  # Lista
print({'Nombre': 'John Doe', 'edad': 30})  # Diccionario
print({1, 2, 3})  # Conjunto
print((4, 5))     # Tupla
```

### 2.2 Variables
Son referencias a objetos.
* Las variables y los objetos se almacenan en diferentes zonas de memoria.
* Las variables siempre referencian a objetos y nunca a otras variables.
* Objetos sí que pueden referenciar a otros objetos (ejemplo: listas).

**Sentencia de asignación:** `<nombre_variable> = <objeto>`

```python
a = 1          # entero
b = 4.0        # coma flotante
c = "ITQ"      # string
d = 10 + 1j    # numero complejo
e = True       # boolean
f = None       # None

print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
print(e, type(e))
print(f, type(f))
```

#### Las variables no tienen tipo
Las variables apuntan a objetos que sí lo tienen. Dado que Python es un lenguaje de **tipado dinámico**, la misma variable puede apuntar, en momentos diferentes de la ejecución, a objetos de diferente tipo.

```python
a = 3
print(a, type(a))

a = 'Pablo García'
print(a, type(a))

a = 4.5
print(a, type(a))
```

> **Garbage collection:** Cuando un objeto deja de estar referenciado, se elimina automáticamente de la memoria.

#### Identificadores y Direcciones de Memoria
Podemos obtener un identificador único (su dirección de memoria) usando `id()`.

```python
a = 3
print(id(a))
a = 'Pablo García'
print(id(a))
```

#### Referencias compartidas
Un mismo objeto puede ser referenciado por más de una variable. 

*Ojo con el comportamiento de caché en enteros pequeños `[-5, 256]` frente a enteros grandes:*

```python
# Enteros pequeños (Comparten id automáticamente)
a = 25
b = 25
print("Id de a(25):", id(a))
print("Id de b(25):", id(b))

# Enteros grandes (No comparten id si se definen por separado)
a = 258
b = 258
print("Id de a(258):", id(a))
print("Id de b(258):", id(b))
```

#### Reglas y convenciones para nombres de variables:
1. No se puede poner números delante del nombre de las variables.
2. Por convención, evitar `CamelCase`. Mejor usar `snake_case` (uso de `_` para separar palabras).
3. El lenguaje diferencia entre mayúsculas y minúsculas (`case-sensitive`).
4. Deben ser descriptivos.
5. Hay palabras o métodos reservados (**Built-ins** y **KeyWords**). ¡Cuidado con reasignarlos!

```python
print(pow(3, 2))

# Ejemplo de peligro reasignando un built-in
pow = 1 
print(pow)
# Descomentar la siguiente línea provocará un TypeError porque 'pow' ahora es un entero y no una función:
# print(pow(3,2))
```

#### Asignación múltiple de variables e intercambio de valores

```python
x, y, z = 1, 2, 3
print(x, y, z)

# Intercambio de valores directo en Python
a = 1
b = 2
a, b = b, a
print("Intercambio:", a, b)
```

## 2.3 Tipos de datos básicos

### Booleanos (`bool`)
Tiene dos posibles valores: `True` o `False`. También se guardan en caché.

```python
a = True
b = True
print(a is b)  # True, apuntan al mismo objeto en caché
print(a == b)
```

### Números
* **Enteros:** Sin parte fraccional.
* **Coma flotante:** Con parte fraccional.
* **Complejos:** Ejemplo `2 + 4j`.
* **Representaciones:** Base 10 (decimal), Base 2 (`0b` binario), Base 8 (`0o` octal), Base 16 (`0x` hexadecimal).

```python
x = 58             # decimal
z = 0b00111010     # binario
w = 0o0072         # octal
y = 0x3A           # hexadecimal

print(x == y == z == w) # Dan el mismo valor decimal
```

### Strings (`str`)
Cadenas de caracteres. 
* Son secuencias (la posición importa).
* Son **inmutables**.

```python
s = 'John "ee" Doe'
print(s[0])       # Primer carácter
print(s[-1])      # Último carácter
print(s[1:8:2])   # Slicing [inicio:fin:paso]
print(s + "e")    # Concatenación
```

## 2.4 Conversión entre tipos (Casting)
Permite obtener objetos de un tipo a partir de otro.

```python
print(int(2.8))
print(float("3"))
print(str(3.0))
print(bool(""))      # Vacío es False
print(bool("a"))     # Con texto es True
print(bool(0))       # 0 es False
print(bool(None))    # None es False
```

#### Tipos de división:
* `/` División convencional (siempre devuelve `float`).
* `//` División entera (truncada, devuelve `int`).

```python
print(7 / 4)   # convencional
print(7 // 4)  # entera
```

### Aritméticos
* `+` Suma, `-` Resta, `*` Multiplicación, `/` División, `//` División entera, `%` Módulo, `**` Exponenciación.

### Comparación
* `>`, `<`, `==`, `!=`, `>=`, `<=` (Devuelven un booleano).

### Lógicos
* `and`, `or`, `not`.

### Identidad y Pertenencia
* `is`, `is not` (Comparan identidad/id).
* `in`, `not in` (Comprueban si un elemento pertenece a una secuencia).

```python
x = 10
lista = [1, 2, 3, 4, 10]
print(x in lista)
print(x is not float)
```

## 3. Tipos de datos compuestos (Colecciones)

### 3.1 Listas
Colección ordenada de objetos, mutables, heterogéneas, de tamaño dinámico y soportan anidamiento.

```python
letras = ['a', 'b', 'c', 'd']
print("Longitud:", len(letras))

# Agregar elementos
letras.append('e')
print("Append:", letras)

# Slicing en listas
print("Slice [1:3]:", letras[1:3])
```

### 3.2 Diccionarios (`dict`)
Mapeos del tipo **Clave-Valor**. Son mutables. Las claves deben ser objetos inmutables (como strings o enteros) y únicas.

```python
persona = {'DNI': '11111111D', 'Nombre': 'Carlos', 'Edad': 34}
print(persona['Nombre'])

# Método seguro get (con valor por defecto si no existe la clave)
print(persona.get('Trabajo', "En el paro"))
```

### 3.3 Sets (Conjuntos)
Colecciones mutables de elementos únicos **no ordenados**. Solo contienen objetos inmutables y no permiten anidamiento directo de estructuras mutables.

```python
conjunto1 = {1, 2, 2, 3, 4, 4} # Los duplicados se eliminan automáticamente
conjunto2 = {3, 4, 5, 6}
print("Unión:", conjunto1 | conjunto2)
print("Intersección:", conjunto1 & conjunto2)
```

### 3.4 Tuplas
Colecciones ordenadas **inmutables**. Se definen mediante el uso de paréntesis `()`.

```python
mi_tupla = ('Foo', 34, 5.0, 34)
print(mi_tupla[0])
# Ojo con la tupla de un solo elemento (lleva coma):
tupla_un_elemento = (1,)
print(type(tupla_un_elemento))
```

#### Ejercicio 5
