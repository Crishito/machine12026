```python
# Creación aleatoria de datos
import numpy as np
import random
random.seed(42)
```

```python
# Creo una lista vacía
nivel_estudios = []

# Agrego los niveles de estudio
nivel_estudios.extend(['Secundaria'] * 110)
nivel_estudios.extend(['Bachillerato'] * 172)
nivel_estudios.extend(['Grado'] * 145)
nivel_estudios.extend(['Master'] * 52)
nivel_estudios.extend(['Doctorado'] * 21)

# Mezclo los elementos de la lista
nivel_estudios = np.random.permutation(nivel_estudios)
```

```python
print(nivel_estudios)
```

```python
# DIAGRAMA DE BARRAS
# Importo las librerías
import matplotlib.pyplot as plt
import numpy as np

# Obtengo los niveles de estudio y su frecuencia
estudios, frecuencia = np.unique(nivel_estudios, return_counts=True)

print(estudios, frecuencia)
```

```python
# Asigno etiquetas al gráfico
plt.ylabel('Frecuencia absoluta')
plt.xlabel('Nivel de estudios')
plt.title('Diagrama de barras')

# Creo el diagrama de barras
plt.bar(estudios, frecuencia) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

# Muestro el gráfico
plt.show()
```

```python
# DIAGRAMA DE SECTORES
# Asigno los colores del gráfico
colores = {'red', 'blue', 'orange', 'green', 'cyan'}

# Creo el diagrama de sectores
plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')

# Asigno un título
plt.title('Diagrama de sectores')

# Muestro el gráfico
plt.show()
```

# Tarea implementar con pandas esta información

```python
# Importo las librerías
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Permutación aleatoria con NumPy
# Mezclo los elementos de la lista
nivel_estudios = np.random.permutation(nivel_estudios)
```

```python
# Convertir la lista permutada en una Serie de Pandas
serie_estudios = pd.Series(nivel_estudios)
```

```python
# Defino el orden de los niveles de estudio
orden_academico = ['Secundaria', 'Bachillerato', 'Grado', 'Master', 'Doctorado']

# Obtengo la frecuencia de cada nivel
frecuencia = serie_estudios.value_counts().reindex(orden_academico)

print(frecuencia)
```

```python

# Creo el diagrama de barras
frecuencia.plot(kind='bar', color='skyblue', edgecolor='black')

# Asigno etiquetas al gráfico
plt.title('Diagrama de Barras con Pandas')
plt.xlabel('Nivel de estudios')
plt.ylabel('Frecuencia absoluta')

# Mantengo las etiquetas horizontales
plt.xticks(rotation=0) # Evita que las etiquetas del eje X se roten de forma incómoda

plt.tight_layout()
plt.show()
```

```python
# Asigno los colores del gráfico
colores_personalizados = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']

# Creo el diagrama de sectores
frecuencia.plot(
    kind='pie',
    autopct='%1.2f%%',
    colors=colores_personalizados,
    startangle=90,
    shadow=False
)

# Asigno un título
plt.title('Diagrama de Sectores con Pandas')

# Oculto la etiqueta del eje Y
plt.ylabel('')

plt.tight_layout()
plt.show()
```
