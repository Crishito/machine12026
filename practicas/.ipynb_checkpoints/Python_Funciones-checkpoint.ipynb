# Estructuras de control

## Selección (sentencias condicionales)

* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:’

````python
# Asigno el bvalo uno a la variable x

x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
````

* Las condiciones son expresiones que se evalúan a *True* o *False.*
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

````python
# Asigno valores a las variables
x = 3
y = [1, 2, 3]

# Verifico diferentes condiciones
print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
````

* Se pueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave *elif.*

````python
# Asigno el valor -3 a x
x = -3

# Verifico si el número es positivo, nulo o negativo
if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
````

* Anidamiento.

````python
# Asigno el valor 120 a val
val = 120

# Verifico el valor de la variable
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')
else:
    print('Valor negativo')
````

# Switch

````python
# Asigno el valor -6 a la variable a
a = -6

# Evalúo el valor de a con match
match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        2
        print('negative')
````

````python
# Asigno el día a la variable a
a = 'l' #week day

# Evalúo el día con match
match a:
    case 'L':
        print("Lunes")
    case 'M':
        print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        print("Wrong Day")
````

**Expresión ternaria**

* Estructura if-else condensado en una línea.
* Se recomienda usar solo en casos sencillos.

````python
 # # # Ejemplo: cálculo del valor absoluto de un número.

x = -3
resultado = x if x >= 0 else -x
print(resultado)
````

````python
# Asigno el valor -3 a val
val = -3

# Obtengo el valor absoluto
if val >=0:
    resultado = val
else:
    resultado = -val
print(resultado)
````

*Concatenación de comparaciones*

  * Se pueden concatenar comparaciones en una misma expresión:
  * *and:* true, si ambos son true.
  * *or:* true, si al menos uno es true.
  * *not:* inversión del valor de verdad de una expresión.

* ‘elif’ para comparar tras un ‘if’

Enlace a [Tablas de Verdad](https://es.wikipedia.org/wiki/Tabla_de_verdad)

````python
# Asigno valores a las variables
genero = 'Drama'
fecha_de_estreno = 1989

# Verifico el género y la fecha de estreno
if genero == 'Comedia' or genero == 'Acción':
    print('Buena película!')
elif fecha_de_estreno >= 1990 and fecha_de_estreno < 2000:
    print('Buena década!')
else:
    print('Meh')
````

**Comparación de variables: ‘==’ vs ‘is’**

* ‘==’ compara si el valor de las variables es el mismo
* ‘is’ compara si los objetos en las variables son iguales (referencia)

````python
# Asigno el valor 1000 a a
a = 1000
b = a

# Comparo las variables
print(a is b)
print(a == b)

c = type(a)
print(c)

d = type(b)
print(d)

print(c == d)

print(id(a))
print(id(b))
````

````python
# Asigno valores a las listas
a = [1, 2, 3]
b = [1, 2, 3]
c = a

# Comparo las listas
print(a is b)
print(a is c)

print(a == b)
print(id(a))
print(id(b))
print(id(c))
````

**Valor booleano**

* Todos los objetos en Python tiene inherentemente un valor booleano: *True* o *False.*
* Cualquier número distinto de 0 ó cualquier objeto no vacío tienen valor *True.*
* El número 0, objetos vacíos y el objeto especial *None* tienen valor *False.*

````python
# Asigno valores a las variables
a = -10  # True
b = None  # False

# Verifico los valores booleanos
if a:
    print("a es True")

if not b:
    print("b es False")
````

````python
# Asigno una lista a c
c = [1,2,3]

# a c le reasigno None
c = None #0 es igual a vacio a nada

if c:
    print("Lista no vacía")
else:
    print("Lista vacía")
````

# Iteración (bucles)

* Repetición de un bloque de código.
* La terminación del bucle depende del tipo de bucle.
* Dos tipos: *while y for.*

**Bucle ‘while’**

* Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta
que una condición evalue a False).
* Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
* Cuidado con los bucles infinitos.

Formato general:

```python
while test:        # Mientras se cumple la condición
    statements     # Instrucciones a ejecutar
```

````python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista

# Asigno una lista y un índice
indice = 0
numeros = [9, 4, 7, 1, 2]

# Muestro los primeros 3 elementos
while indice < 3:
    print(numeros[indice])
    indice = indice + 1
````

````python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0  # Cuenta los menores a 10
indice = 0    # Recorre la lista

while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1

print('Contador:', contador)
print('Indice:', indice)
````
