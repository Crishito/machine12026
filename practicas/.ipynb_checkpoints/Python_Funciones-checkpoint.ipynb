# Funciones

## ¿Qué son?

* Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera que se puedan ejecutar cuando el programador considere oportuno.
* Las funciones van identificadas por un nombre.
* Se debe usar este nombre para invocar (ejecutar) el código de la función.
* Las funciones calculan un resultado (output) en base a unos parámetros de entrada (input).
* En cada invocación de la función, los parámetros de entrada pueden variar.

## Ejemplo

* Las condiciones son expresiones que se evalúan a *True* o *False.*
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

```python
# Creo una función que suma dos valores
def sumar(x, y):
    suma = x + y
    return suma  # Devuelvo el resultado de la suma

# Llamo a la función con diferentes valores
print(sumar(4, 5))
print(sumar(-2, 0))
```

```python
# Convierto una temperatura de grados Celsius a Fahrenheit
#f = c * 1.793 + 32

c = 30
print(c * 1.793 + 32) # celsius to fahrenheit

# Realizo la conversión con otro valor de temperatura
c = 26
print(c * 1.793 + 32)
```

```python
# Creo una función para convertir grados Celsius a Fahrenheit
def convert_celsius_to_fahrenheit(c):
    return c * 1.791117 + 32  # Devuelvo la temperatura convertida
```

```python
# Defino la temperatura en grados Celsius
degrees = 15

# Llamo a la función y muestro la temperatura convertida
print(convert_celsius_to_fahrenheit(degrees))
```

## ¿Qué ventajas nos proporcionan?

* Eliminación de duplicación de código.
* Incremento de la reutilización de código.
* Manejo de complejidad: permiten descomponer grandes sistemas en piezas más pequeñas y, por tanto, más manejables y comprensibles.

## Programación de funciones en Python

Las funciones se definen por medio de la palabra clave *def.*

Formato general:

    def name(arg1, arg2, ..., argN):

        statements
    
Muy comúnmente:

    def name(arg1, arg2, ..., argN):

        ...

        return value

* Las funciones son instrucciones convencionales, que se ejecutan cuando el flujo de control
llega a ellas.
* La definición de la función debe ejecutarse antes de poder invocarse la función

```python
# Defino una función que une las claves con sus valores
def crear_diccionario(keys, values):
    return dict(zip(keys, values))
```

```python
# Después, invoco la función con nombres y edades
nombres = ['Alfred', 'James', 'Peter', 'Harvey']
edades = [87, 43, 19, 16]

# Creo el diccionario usando la función definida anteriormente
usuarios = crear_diccionario(nombres, edades)

# Muestro el diccionario resultante
print(usuarios)
```

* Las funciones pueden estar anidadas en sentencias condicionales o bucles.

```python
# Defino el valor que determina la operación
a = -2

# Si a es positivo, la función suma los valores
if a > 0:
    def operacion(x, y):
        return x + y

# Si a no es positivo, la función multiplica los valores
else:
    def operacion(x, y):
        return x * y

# Llamo a la función y muestro el resultado
print(operacion(3, 4))
```

* Argumentos y return son opcionales.

```python
# Defino una función que muestra un mensaje
def print_hola_mundo():
    print('Hola Mundo')
    return  # Es opcional, pero indica el final de la función

# Llamo a la función
print_hola_mundo()
```

* Cuando una función no tiene sentencia return, la función devuelve el objeto None.

```python
# Defino una función que muestra un mensaje
def print_hola_mundo():
    print('Hola Mundo')
    # return None

# Guardo en una variable el valor retornado por la función
valor = print_hola_mundo()

# Muestro el valor retornado, que en este caso es None
print(valor)
```

* La sentencia *return* puede aparecer en cualquier parte de la función.

```python
# Defino una función que evita la división entre cero
def div_safe(x, y):
    if y == 0:
        return "hola"  # Devuelvo un texto si el divisor es cero

    return x / y  # Realizo la división si no existe error

# Muestro una división válida
print(div_safe(6, 2))

# Muestro el resultado cuando el divisor es cero
print(div_safe(6, 0))

# Guardo el resultado y muestro su tipo de dato
a = div_safe(6, 0)
print(type(a))
```

```python
# Defino una función que realiza una división segura
def div_safe(x, y):
    if y != 0:
        return x / y  # Divide cuando el divisor es diferente de cero
    else:
        return 0  # Devuelve cero para evitar la división entre cero

# Llamo a la función y muestro el resultado
print(div_safe(6, 0))
```

* Puede haber más de un valor de retorno. Se devuelve una tupla

```python
# Defino una función que devuelve los tres primeros elementos
def devolver_primero_segundo(coleccion):
    return coleccion[0], coleccion[1], coleccion[2]

# Llamo a la función y guardo los valores retornados
t = devolver_primero_segundo(['Alfred', 'James', 'Peter', 'Harvey'])

# Muestro la tupla completa, su segundo elemento y su tipo de dato
print(t)
print(t[1])
print(type(t))
```

```python
# Guardo por separado los tres valores que devuelve la función
primero, segundo, tercero = devolver_primero_segundo(
    ['Alfred', 'James', 'Peter', 'Harvey']
)

# Muestro cada valor de forma individual
print(primero)
print(segundo)
print(tercero)
```

* Los argumentos pueden tener valores por defecto, siempre al final.

```python
# Defino una función con tres parámetros, donde c es opcional
def f(a, b, c=None):
    if a is not None:
        print('Se ha proporcionado a')

    if b is not None:
        print('Se ha proporcionado b')

    if c is not None:
        print('Se ha proporcionado c')

# Llamo a la función proporcionando los tres valores
f(1, 5, 2)
```

* Los parámetros se pasan por posición, pero el orden se puede alterar si se especifica el nombre del parámetro en la llamada, named values.

```python
# Defino una función que calcula una potencia
def power(base, exponente):
    return pow(base, exponente)

# Llamo a la función usando argumentos por posición
print(power(2, 3))

# Llamo a la función usando argumentos por nombre
print(power(exponente=3, base=2))
```

• Al ejecutarse una sentencia *def,* se crea un **objeto de tipo función** y éste se asocia con el nombre especificado para la función.
• Un objeto de tipo función es como cualquier otro tipo de objeto.
• Variables pueden referenciar objetos de tipo función.

```python
# Defino una función que suma dos valores
def sumar(x, y):
    return x + y

# Asigno la función sumar a la variable f
f = sumar

# Muestro que ambas variables apuntan a la misma función
print(id(f))
print(id(sumar))

# Muestro el tipo de dato de f
print(type(f))

# Llamo a la misma función usando ambos nombres
print(sumar(5, 6))
print(f(5, 6))
```

```python
# Defino una función que aplica varias operaciones a cada texto
def clean_strings(strings, ops):
    result = []

    # Recorro cada texto de la lista
    for value in strings:
        # Aplico las funciones indicadas en el orden establecido
        for function in ops:
            value = function(value)

        result.append(value)

    return result

# Defino una función que agrega "_5" al final del texto
def concat5(x):
    return x + '_5'

# Creo una lista de textos con espacios
strings = [' valencia ', ' barcelona', ' bilbao ']
print(strings)

# Guardo las funciones que se aplicarán a cada texto
operations = [str.strip, str.title, concat5]

# Muestro la lista limpia y modificada
print(clean_strings(strings, operations))
```

## Paso de parámetros

* Tipos simples (inmutables) por **valor:** int, float, string.
  * El efecto es como si se creara una copia dentro de la función, aunque realmente no es esto lo que ocurre.
  * Los cambios dentro de la función no afectan fuera.

```python
# Defino una función que aumenta en 3 el valor recibido
def cambiar_valor(x):
    x += 3
    print('Dentro de la función:', x)  # El cambio solo afecta a la variable local
    return

# Defino el valor inicial
a = 2

# Llamo a la función con el valor de a
cambiar_valor(a)

# Muestro que el valor original de a no cambia
print('Fuera de la función:', a)
```

* Tipos complejos (mutables) por referencia: list, set, dictionary.
  * Dentro de la función se maneja el mismo objeto que se ha pasado desde fuera.
  * Los cambios dentro de la función sí afectan fuera.

```python
# Defino una función que agrega el número 4 a una lista
def anyadir_2_a(x):
    x.append(4)

# Creo una lista y la envío a la función
lista = [0, 1, 3]
anyadir_2_a(lista)

# Muestro la lista modificada
print(lista)
```

```python
# Defino una función que muestra la identidad del objeto recibido
def anyadir_2_a(x):
    print(id(x))
    return

# Creo una tupla y muestro su identidad
lista = (0, 1)
print(id(lista))

# Envío la tupla a la función
anyadir_2_a(lista)

# Muestro que la tupla permanece sin cambios
print(lista)
```

* Si quiero modificar un objeto que es de un tipo básico, los cuales se pasan por valor, puedo simplemente devolver el resultado de la función y hacer una asignación:

```python
# Defino una función que multiplica por 2 el valor recibido
def multiplicar_por_2(x):
    x = x * 2
    return x

# Defino el valor inicial
a = 3

# Guardo en a el resultado devuelto por la función
a = multiplicar_por_2(a)

# Muestro el nuevo valor de a
print(a)
```

* Si quiero que no se modifique un objeto de un tipo complejo, los cuales se pasan por referencia, puedo pasar una copia a la función:

```python
# Defino una función que agrega el número 2 a una lista
def anyadir_2_a(x):
    x.append(2)
    print('Dentro de la función: ', x)

# Creo una lista con dos elementos
lista = [0, 1]

# Envío una copia para no modificar la lista original
anyadir_2_a(lista.copy())

# Muestro que la lista original permanece igual
print('Fuera de la función: ', lista)
```

```python
# Alternativa para obtener una copia mediante slicing
lista = [0, 1]

# Envío una copia de la lista para no modificar la original
anyadir_2_a(lista[:])

# Muestro que la lista original permanece igual
print('Fuera de la función: ', lista)
```

```python
# Importo el módulo para realizar una copia profunda
import copy

# Creo una lista con elementos simples y listas anidadas
lista = [2, 4, 16, 32, [34, 10, [5, 5]]]

# Creo una copia independiente de todos los elementos
copia = copy.deepcopy(lista)

# Modifico valores de la copia
copia[0] = 454
copia[4][2][0] = 64

# Muestro la lista original y la copia modificada
print(lista)
print(copia)

# Comparo las identidades de ambas listas
print(f"{id(lista)} - {id(copia)}")
print(f"{id(lista[4])} - {id(copia[4])}")
```

* Reasignar un parámetro nunca afecta al objeto de fuera:

```python
# Defino una función que crea una nueva lista dentro de ella
def funcion_poco_util(x):
    print(id(x))

    # x deja de apuntar a la lista original y referencia una nueva lista
    x = [2, 3]
    x.append(4)

    print(id(x))

# Creo una lista y muestro su identidad
lista = [0, 1]
print(id(lista))

# Envío la lista a la función
funcion_poco_util(lista)

# Muestro que la lista original no fue modificada
print(lista)
```

```python
# Defino una función que muestra la identidad de la tupla recibida
def modif_tupla(a):
    print(id(a))

# Creo una tupla y muestro su identidad
t = (1, 2)
print(id(t))

# Envío la tupla a la función
modif_tupla(t)

# Muestro que la tupla permanece sin cambios
print(t)
```

## Argumentos arbitrarios

* No se sabe a priori cuantos elementos se reciben.
* Se reciben como una tupla.
* Los parámetros se especifican con el símbolo ’*’.

```python
# Defino una función que puede recibir varios nombres
def saludar(*names):
    print(type(names))  # Los argumentos se guardan en una tupla

    # Recorro y saludo a cada nombre recibido
    for name in names:
        print("Hola " + name)

# Llamo a la función sin argumentos
saludar()

# Llamo a la función con varios nombres
saludar("Manolo", "Pepe", "Luis", "Alex", "Juan")
```

```python
# Defino una función que recibe varios argumentos con nombre
def consulta_sql(**kwargs):
    print(kwargs)  # Los argumentos se guardan en un diccionario

    # Creo la instrucción SQL inicial
    sql = "SELECT * FROM bicis WHERE "

    # Convierto cada clave y valor en una condición
    args = [f"{key}={value}" for key, value in kwargs.items()]

    # Uno las condiciones mediante AND
    sql += " AND ".join(args)

    return sql

# Llamo a la función y muestro la consulta creada
print(consulta_sql(station='alameda', free=10))
```

## Recursividad

* Funciones que se llaman a sí mismas

```python
# Defino una función recursiva para calcular el factorial
def factorial(n):
    if n == 0:
        return 1  # Caso base de la recursividad
    else:
        return n * factorial(n - 1)

# Muestro el factorial de 10
print(factorial(10))


# Defino otra función para calcular el factorial con un ciclo
def fact(n):
    a = 1

    # Multiplico los números desde 1 hasta n
    for i in range(1, n + 1):
        a *= i

    return a

# Muestro el factorial de 10
print(fact(10))
```

```python
# Defino una función lambda recursiva para calcular el factorial
fact = lambda x: 1 if x == 0 else x * fact(x - 1)

# Calculo el factorial de 10
fact(10)
```

## Funciones Lambda

* Es posible definir funciones anónimas (sin nombre).
* Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia *def* no puede (por ejemplo, dentro de una lista o como parámetro de una función).
* Útiles cuando se quiere pasar una función como argumento a otra.
 
Formato general:

`lambda <lista de argumentos> : <valor a retornar>`

* Importante: *<valor a retornar>* no es un conjunto de instrucciones. Es simplemente una expresión *return* que omite esta palabra.
* Las funciones definidas con *def* son más generales:
  * Cualquier cosa que implementes en un lambda, lo puedes implementar como una función convencional con *def,* pero no viceversa.

```python
# Ejemplo: función suma.
# Tres valores
def suma(x, y, z):
    return x + y + z

print(suma(1, 2, 3))


# # Ejemplo: función suma como expresión lambda.
suma2 = lambda x, y, z: x + y + z

print(suma2(1, 2, 3))
```

```python
# La funcion 'sort' puede recibir una función optional como parámetro.
# Esta función 'key' se invoca para cada elemento de la lista antes de realizar las comparaciones.
# Ejemplo: ordenar strings por número de carácteres.
    
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

cities.sort()
print(cities)

cities.sort(key = lambda x : len(x))
print(cities)

cities.sort(key = len)
print(cities)

def longitud(x):
    return len(x)
    
cities.sort(key = longitud)
print(cities)
```

```python
# Mismo ejemplo sin lambda

cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

def num_caracteres(x):
    return len(x)

cities.sort(key = num_caracteres)
print(cities)
```

* Nos podemos guardar una referencia para utilización repetida.

```python
# Creo una lista con nombres de ciudades
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

# Defino una función lambda que obtiene la longitud de cada ciudad
f = lambda x: len(x)

# Recorro las ciudades y muestro la cantidad de letras
for s in cities:
    print(f(s))
```

* Las expresiones lambda pueden formar parte de una lista

```python
# Creo una lista de funciones lambda para calcular potencias
pows = [
    lambda x: x ** 0,
    lambda x: x ** 1,
    lambda x: x ** 2,
    lambda x: x ** 3,
    lambda x: x ** 4
]

# Recorro las funciones y calculo las potencias de 2
for f in pows:
    print(f(2))
```

* Diccionarios se pueden utilizar como ‘switch’ en otros lenguajes
* Ideal para escoger entre varias opciones

```python
# Defino una función que selecciona una operación mediante un diccionario
def switch_dict(operator, x, y):
    # Creo las opciones disponibles y ejecuto la seleccionada
    return {
        'add': lambda: x + y,
        'sub': lambda: x - y,
        'mul': lambda: x * y,
        'div': lambda: x / y,
    }.get(operator, lambda: None)()  # Devuelve None si la operación no existe

# Pruebo la función con diferentes operaciones
print(switch_dict('add', 2, 2))
print(switch_dict('div', 10, 2))
print(switch_dict('mod', 17, 3))
```

## Scopes

* La localización de una asignación a una variable en el código determina desde donde puedes acceder a esa variable.
* Por ejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.
* El alcance de la visibilidad de una variable determina su *scope.*
* El término *scope* hace referencia a un espacio de nombres *(namespace).* Por ejemplo, una
función establece su propio *namespace.*

```python
# Defino una variable global
X = 10

# Creo una función con una variable local diferente
def func():
    X = 20  # Esta variable solo existe dentro de func

    # Creo una función interna con su propia variable local
    def func_int():
        X = 30
        print(X)

    # Llamo a la función interna
    func_int()

# Llamo a la función principal
func()
```

**Regla LEGB** Dada una función *F,* tenemos los siguientes *scopes:*

* **L**ocal: variables locales a *F.*
12
* **E**nclosing: variables localizadas en funciones que contienen a (o por encima de) F.
* **G**lobal: variables definidas en el módulo (fichero) que no están contenidas en ninguna función.
    Este scope no abarca más de un fichero.
* **B**uilt-ins: proporcionadas por el lenguaje.

La resolución de nombres ocurre de abajo a arriba.

```python
# Global (X y func)
# Defino una variable global
X = 99

# Creo una función que recibe una variable local
def func(Y):
    Z = X + Y  # Sumo la variable global X con la variable local Y
    return Z

# Llamo a la función y muestro el resultado
print(func(1))
```

```python
# Defino una variable global
X = 1

# Creo una función con una variable local diferente
def func():
    X = 2  # Solo cambia dentro de la función

# Llamo a la función
func()

# Muestro la variable global, que conserva su valor
print(X)
```

* La sentencia *global* nos permite modificar una variable global desde dentro de una función.

```python
# Defino una variable global
X = 1

# Creo una función que modifica la variable global
def func():
    global X
    X = 2

# Llamo a la función
func()

# Muestro el nuevo valor de la variable global
print(X)
```

• No hay necesidad de usar global para referenciar variables. Sólo es necesario para modificarlas.


```python
# Defino dos variables globales
y, z = 1, 2

# Creo una función que genera otra variable global
def todas_globales():
    global x
    x = y + z  # Sumo las variables globales y y z

# Llamo a la función
todas_globales()

# Muestro los valores de las variables globales
print(x)
print(y)
print(z)
```

```python
# Defino una variable global
X = 77

# Creo una función con una variable local que será usada por la función interna
def f1():
    X = 88  # Variable del ámbito envolvente para f2

    # La función interna accede al valor de X definido en f1
    def f2():
        print(X)

    # Llamo a la función interna
    f2()

# Llamo a la función principal
f1()
```

**Closures**

* Las funciones pueden recordar su enclosing *scope,* independientemente de si éstos continuan existiendo o no

```python
# Defino una función principal con una variable local
def f1():
    X = 88  # Variable del ámbito envolvente para f2

    # Creo una función interna que muestra el valor de X
    def f2():
        print(X)

    # Devuelvo la función f2 sin ejecutarla
    return f2

# Guardo en accion la función devuelta por f1
accion = f1()

# Ejecuto la función interna
accion()
```

* Sentencia *nonlocal* para modificar variables que no son locales, pero tampoco globales.

```python
# Defino una función principal con una variable contador
def f1():
    contador = 10

    # Creo una función interna que modifica el contador
    def f2():
        nonlocal contador
        contador += 1
        print(contador)

    # Devuelvo la función interna
    return f2

# Guardo la función devuelta en la variable accion
accion = f1()

# Ejecuto la función varias veces
accion()
accion()
accion()
```

```python
# Ejercicio 1
# Defino una función que calcula el cuadrado de cada número
def calcular_cuadrados(numeros):
    cuadrados = []

    # Recorro la lista y agrego cada número elevado al cuadrado
    for numero in numeros:
        cuadrados.append(numero ** 2)

    return cuadrados

# Llamo a la función y muestro el resultado
lista = [1, 2, 3, 4, 5]
print(calcular_cuadrados(lista))
```

```python
# Ejercicio 2
# Defino una función que puede recibir una cantidad indeterminada de números
def sumar_numeros(*numeros):
    suma = 0

    # Recorro los números y acumulo sus valores
    for numero in numeros:
        suma += numero

    return suma

# Llamo a la función con diferentes cantidades de números
print(sumar_numeros(2, 4, 6))
print(sumar_numeros(1, 3, 5, 7, 9))
```

```python
# Ejercicio 3
# Defino una función que invierte el texto recibido
def invertir_texto(texto):
    return texto[::-1]

# Llamo a la función y muestro el resultado
print(invertir_texto("hola"))
```

```python
# Ejercicio 4
# Creo una función lambda que invierte el texto
invertir_texto = lambda texto: texto[::-1]

# Llamo a la función lambda
print(invertir_texto("hola"))
```

```python
# Ejercicio 5
# Defino una función que comprueba si un número está dentro del rango
def comprobar_rango(numero, inicio, fin):
    return inicio <= numero <= fin

# Compruebo si el número 5 está entre 1 y 10
print(comprobar_rango(5, 1, 10))

# Compruebo si el número 15 está entre 1 y 10
print(comprobar_rango(15, 1, 10))
```

```python
# Ejercicio 6
# Defino una función que obtiene los cinco primeros múltiplos
def primeros_multiplos(numero):
    multiplos = []

    # Verifico que el valor sea un entero positivo
    if type(numero) != int or numero <= 0:
        print("Error: debe ingresar un número entero positivo")
        return multiplos

    # Calculo los cinco primeros múltiplos
    for i in range(1, 6):
        multiplos.append(numero * i)

    return multiplos

# Llamo a la función con un valor correcto
print(primeros_multiplos(3))

# Llamo a la función con un valor incorrecto
print(primeros_multiplos(-2))
```

```python
# Ejercicio 7
# Defino una función que comprueba si existen elementos repetidos
def tiene_duplicados(lista):
    elementos = []

    # Recorro los elementos de la lista
    for elemento in lista:
        if elemento in elementos:
            return True

        elementos.append(elemento)

    return False

# Pruebo la función con diferentes listas
print(tiene_duplicados([1, 2, 3, 2]))
print(tiene_duplicados([1, 2, 3, 4]))
```

```python
# Ejercicio 8
# Creo una función lambda que compara la longitud de la lista y del conjunto
tiene_duplicados = lambda lista: len(lista) != len(set(lista))

# Pruebo la función con diferentes listas
print(tiene_duplicados([1, 2, 3, 2]))
print(tiene_duplicados([1, 2, 3, 4]))
```

```python
# Ejercicio 9
# Defino una función que comprueba si el texto es un palíndromo
def es_palindromo(texto):
    # Convierto el texto a minúsculas y elimino los espacios
    texto = texto.lower().replace(" ", "")

    # Comparo el texto original con el texto invertido
    return texto == texto[::-1]

# Pruebo la función con diferentes palabras
print(es_palindromo("reconocer"))
print(es_palindromo("python"))
```
