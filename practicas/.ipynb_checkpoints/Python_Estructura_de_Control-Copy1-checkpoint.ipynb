````python
# Asigno el bvalo uno a la variable x

x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
````

# Switch

````python
# Asigno el valor -6 a la variable a
a = -6

# Evalúo el valor de a con match
match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        2
        print('negative')
````

````python
# Asigno el día a la variable a
a = 'l' #week day

# Evalúo el día con match
match a:
    case 'L':
        print("Lunes")
    case 'M':
        print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        print("Wrong Day")
````

# Iteración (bucles)

* Repetición de un bloque de código.
* La terminación del bucle depende del tipo de bucle.
* Dos tipos: *while y for.*

**Bucle ‘while’**

* Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta
que una condición evalue a False).
* Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
* Cuidado con los bucles infinitos.

Formato general:

```python
while test:        # Mientras se cumple la condición
    statements     # Instrucciones a ejecutar
```

````python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista

# Asigno una lista y un índice
indice = 0
numeros = [9, 4, 7, 1, 2]

# Muestro los primeros 3 elementos
while indice < 3:
    print(numeros[indice])
    indice = indice + 1
````

````python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0  # Cuenta los menores a 10
indice = 0    # Recorre la lista

while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1

print('Contador:', contador)
print('Indice:', indice)
````

````python
# Asigno un nombre a la variable
nombre = 'Pablo'

while nombre:  # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]  # Elimino la primera letra
````

````python
# Bucle que muestra los números del 0 al 9
i = 0

while i < 10:
    print(i)
    i += 1  # Aumento el valor de i para evitar un bucle infinito
    
````

**Bucle ‘for’**

* Permite recorrer los ítems de una *secuencia* o un objeto *iterable.*
* Funciona en strings, listas, tuplas, etc.

Formato general:

```python
for item in objeto:    # Asigna los ítems del objeto a la variable item en cada iteración
    statements         # Instrucciones a ejecutar
```

````python
# Creo una lista de películas
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']

# Recorro y muestro cada película de la lista
for pelicula in peliculas:
    print(pelicula)
````

* También se usa para iterar un número preestablecido de veces *(counted loops):*

````python
# Recorro y muestro los números del 0 al 9
for i in range(10):
    print(i)
````

* *range* es útil en combinación con *len* porque permite acceder a los elementos de una *secuencia* por posición.


````python
# Recorro y muestro cada letra del nombre
nombre = 'Pablo'

for i in range(len(nombre)):
    print(nombre[i])
````

````python
# Recorro y muestro cada letra del nombre
for letra in nombre:
    print(letra)
````

* Iteración de tuplas:

````python
# Recorro las tuplas y muestro sus tres valores
tuplas = [(1, 2, 4), (3, 4), (5, 6)]

for a, b, c in tuplas:
    print(a, b, c)
````

* Iteración de diccionarios. Se iteran las claves:

````python
# Recorro el diccionario y muestro cada clave con su valor
diccionario = {'a': 1, 'b': 2, 'c': 3}

for key in diccionario:
    print(f"{key} => {diccionario[key]}")
````

* Para iterar los pares (clave-valor) o únicamente los valores, se deben usar los métodos *items* y *values.*

````python
# Recorro el diccionario y muestro cada clave con su valor
for key, value in diccionario.items():
    print(f"{key} => {value}")
````

````python
# Recorro y muestro los valores del diccionario
for value in diccionario.values():
    print(value)
````

* La variable *item* en la cabecera del *for* puede ser cualquier expresión que sea válida como parte izquierda de una asignación convencional.

````python
# Recorro las tuplas y muestro sus tres valores
for a, b, c in [(1, 2, 3), (4, 5, 6)]:
    print(a, b, c)
````

## Las sentencias break, continue y else

**Break y continue**

* Solo tienen sentido dentro de bucles.
* *Break* permite terminar el bucle por completo.
* *Continue* permite saltar a la siguiente iteración, continuando con el bucle.
* Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (if).

````python
# Ejemplo: buscar una calificación y detener el ciclo al encontrarla
rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]

for rating in movie_ratings:
    print(rating)

    if rating == rating_to_find:
        print("Found")
        break

print(rating)
````

````python
# Recorro los números del 0 al 9 y muestro únicamente los impares
for i in range(10):
    if i % 2 == 0:
        continue  # Si el número es par, salta a la siguiente iteración

    print(f'Odd number {i}')
````

````python
# Sin continue, la condición se coloca dentro del if
for i in range(10):
    if i % 2 != 0:
        print(f'Odd number {i}')
````

````python
# Recorro dos ciclos y muestro la combinación de sus valores
for i in range(5):
    for a in range(2):
        print(f"{i} - {a}")

        # Detengo únicamente el ciclo interno cuando i es 3 y a es 0
        if i == 3 and a == 0:
            break
````

**Else**
  
* Los bucles pueden tener una sensencia *else.*
* Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros
lenguajes.
* Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de
un *break.*

````python
# Lista de números y elemento que se desea encontrar
lista = [60, 60, 30, 60]
element_to_find = 60

# Recorro la lista para buscar el elemento
for element in lista:
    if element == element_to_find:
        print("found")
        break  # Detengo el ciclo cuando encuentro el elemento
else:
    print("not found")  # Se muestra si el elemento no fue encontrado
````

````python
# Ejemplo: búsqueda de un elemento utilizando una bandera
lista = [10, 30, 50, 40]
element_to_find = 30
found = False  # Indica si el elemento fue encontrado

# Recorro la lista para buscar el elemento
for element in lista:
    if element == element_to_find:
        found = True
        break  # Detengo el ciclo al encontrarlo

# Verifico el valor de la bandera
if found:
    print("dato encontrado")
else:
    print("not found")
````

# List Comprehensions

* Permiten construir listas a través de la ejecución repetida, mediante una sentencia *for,* de una expresión para cada *item* elemento de un objeto *iterable.*

* Van entre ‘[’ y ’]’. Esto es indicativo de que estamos construyendo una lista

Sintaxis:

`[<expresion> for <item> in <iterable>]`

Ejemplo: repetir los caracteres de un string.

````python
# Repito tres veces cada carácter de la palabra Enrique
[char * 3 for char in "Enrique"]
````

````python
# Creo una lista vacía para guardar los caracteres
lista = []

# Recorro cada carácter de la palabra Enrique
for char in 'Enrique':
    lista.append(char)

# Muestro la lista creada
print(lista)
````

Comúnmente, el item de la sentencia *for* aparecerá en la expresión principal, pero eso no es obligatorio.


````python
# Creo una lista con el número 2 por cada carácter de la palabra Enrique
[2 for _ in 'Enrique']
````

### Versión extendida

Se puede especificar un filtro (sentencia *if*) para obtener únicamente los elementos que cumplan cierta condición.

Sintaxis:

`[<expression> for <item> in <iterable> if <condition>]`

Ejemplo: obtener números pares:

````python
# Creo una lista con los números pares del 0 al 8
[y for y in range(9) if y % 2 == 0]
````

* Las list comprehension no son realmente requeridas, ya que siempre podemos escribir un bucle
equivalente.


````python
# Creo una lista vacía para guardar los números pares
pares = []

# Recorro los números del 0 al 8
for x in range(9):
    if x % 2 == 0:
        pares.append(x)  # Agrego el número si es par

# Muestro la lista de números pares
print(pares)
````

### Versión completa

La sentencia *if* de una list comprehension también puede contener una expresión alternativa.

Sintaxis:

`[<expression_1> if <condition> else <expression_2> for <item> in <iterable>]`

Ejemplo: poner a cero los números pares.

````python
# Reemplazo los números pares por 0 y mantengo los impares
[0 if x % 2 == 0 else x for x in range(9)]
````

````python
# Tomo solo los números positivos y elevo al cuadrado los mayores que 2
[x**2 if x > 2 else x for x in range(-4, 5) if x > 0]
````

````python
# Creo una lista vacía para guardar los resultados
lista = []

# Recorro los números desde -4 hasta 4
for x in range(-4, 5):
    if x > 0:  # Tomo únicamente los números positivos
        if x > 2:
            lista.append(x**2)  # Elevo al cuadrado los mayores que 2
        else:
            lista.append(x)  # Mantengo los valores 1 y 2

# Muestro la lista resultante
print(lista)
````

````python
# Elevo al cuadrado los positivos, reemplazo el cero por 100
# y mantengo los números negativos
[x**2 if x > 0 else 100 if x == 0 else x for x in range(-4, 5)]
````

### Anidamiento

Las list comprehensions soportan anidamiento en sus expresiones.
    
Ejemplo: bucle anidado para obtener una lista de tuplas que combinan los elementos de dos listas
dadas.

````python
# Creo dos listas con diferentes valores
lista_1 = [1, 2, 3]
lista_2 = [4, 5]

# Combino cada elemento de la primera lista con cada elemento de la segunda
[(x, y) for x in lista_1 for y in lista_2]
````

### Pros y contras

* Principales ventajas de las comprehensions en comparación a un bucle convencional:
  * Expresión compacta y legible, si estás familiarizado con la sintaxis.
  * Mejor rendimiento.
* Desventaja: no escalan bien. Una list comprehension se puede convertir rápidamente en una
expresión difícil de entender.

### Otros tipos de comprehensions

**Dictionary comprehensions**

Usando ‘{’ y ‘}’ como delimitadores y una expresión ‘clave : valor’, se obtiene un diccionario en lugar de una lista.

Ejemplo: diccionario donde cada valor es el cuadrado de la clave.

````python
# Creo un diccionario donde cada número es la clave y su cuadrado es el valor
d = {x: x * x for x in range(10)}

# Muestro el diccionario creado
print(d)

# Muestro el tipo de dato de la variable
print(type(d))
````

````python
# Creo las listas de productos y precios
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]

# Uno cada producto con su precio y creo un diccionario
shopping = {k: v for k, v in zip(items, price)}

# Muestro el diccionario creado
print(shopping)
````

````python
# Creo las listas de productos y precios
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]

# Uno cada producto con su precio y recorro las tuplas creadas
for k in zip(items, price):
    print(type(k))  # Muestro el tipo de dato
    print(k)  # Muestro el producto junto con su precio
````

**Set comprehensions**

Usando ‘{’ y ‘}’ como delimitadores y una expresión simple (al igual que en las list comprehensions), se obtiene un conjunto.

Ejemplo: conjunto que incluye los 10 primeros números naturales.

````python
# Creo un conjunto con los números del 0 al 9
c = {x for x in range(10)}

# Muestro el conjunto creado
print(c)

# Muestro el tipo de dato de la variable
print(type(c))
````

# Excepciones

* Las excepciones son eventos que representan situaciones excepcionales.
* Alteran el flujo de ejecución convencional.
* Python lanza excepciones automáticamente cuando se producen errores.
* El programador puede lanzar excepciones de manera explícita y también capturar excepciones
para actuar como se crea conveniente.


**Try/except**

* Permite capturar excepciones y actuar en consecuencia.

Ejemplo: error de acceso fuera de rango.

````python
# Creo una lista con cuatro elementos
lista = [6, 1, 0, 5]

# Intento acceder a la posición 4, que no existe
lista[4]

# Esta línea no se ejecuta debido al error anterior
print('Código tras el error')
````

````python
# Creo una lista y defino el índice que deseo consultar
lista = [6, 1, 0, 5]
i = 300

# Intento acceder al elemento de la lista
try:
    a = lista[i]

# Capturo el error si el índice no existe
except IndexError:
    print('He capturado la excepción de tipo IndexError')
    a = 0

# El programa continúa después de manejar el error
print('Código tras el bloque try')
print(a)
````

* Normalmente, al capturar excepciones, querremos ser lo más específicos posible, pero también se puedes usar una sentencia *try-except* que capture cualquier error.

````python
# Intento realizar una división entre cero
try:
    4 / 0

# Capturo el error generado por la división
except:
    print('He capturado la división por cero')
````

* Recoger la excepción e imprimir el mensaje de error

````python
# Creo una lista con cuatro elementos
lista = [6, 1, 0, 5]

# Intento acceder a una posición que no existe
try:
    print(lista[4])

# Capturo la excepción y muestro el mensaje del error
except Exception as e:
    print("Error = " + str(e))

# El programa continúa después de manejar el error
print('Reacheable Code')
````

**Try/finally**

* A través de *finally* podemos especificar código que queremos que se ejecute siempre (independientemente de si se produce la excepción o no).
* Se suele usar para liberar recursos

````python
# Creo una lista con cuatro elementos
lista = [6, 1, 0, 5]

# Intento acceder a una posición válida de la lista
try:
    a = lista[1]

# Capturo el error si el índice no existe
except IndexError:
    print('Exception IndexError Captured')

# Este bloque siempre se ejecuta, exista o no un error
finally:
    print('Bloque finally')

# Accedo a otro elemento de la lista
b = lista[2]

# Muestro que el programa continúa después del bloque try
print('Código tras el bloque try')
print(b)
````

**raise**

* La sentencia raise nos permite lanzar excepciones de manera explícita.

````python
# Creo una lista y defino el índice que deseo validar
lista = [6, 1, 0, 5]
indice = 4

# Verifico si el índice está fuera del rango de la lista
try:
    if indice >= len(lista):
        raise IndexError('Índice fuera de rango')

# Capturo la excepción generada manualmente
except IndexError:
    print('Excepción de tipo IndexError capturada')
````

````python
# Ejercicio 1
#Sumar todos los números de una lista
numeros = [1, 5, 9, 2, 3]
suma = 0

# Recorro la lista y acumulo cada valor
for numero in numeros:
    suma += numero

# Muestro la suma total
print(suma)
````

````python
# Ejercicio 2
# Elimino los números repetidos de la lista
numeros = [1, 2, 2, 3, 4, 5, 4, 5, 3, 3, 1]
resultado = []

# Agrego cada número solo si todavía no está en el resultado
for numero in numeros:
    if numero not in resultado:
        resultado.append(numero)

# Muestro la lista sin elementos repetidos
print(resultado)
````

````python
# Ejercicio 3
# Defino la cantidad de elementos y creo un diccionario vacío
n = 5
diccionario = {}

# Guardo cada número como clave y su cuadrado como valor
for i in range(1, n + 1):
    diccionario[i] = i * i

# Muestro el diccionario creado
print(diccionario)
````

````python
# Ejercicio 4
# Creo una lista de palabras para realizar la búsqueda
palabras = ["", "Valencia", "python", "asignaturas", "programación",
            "java", "estudiantes", "académicos"]

# Busco una palabra que tenga 9 o más letras y empiece con "a"
for palabra in palabras:
    if len(palabra) >= 9 and palabra[0] == 'a':
        print(f"Palabra encontrada: {palabra}")
        break  # Detengo el ciclo al encontrarla
else:
    print("Palabra no encontrada")
````

````python
# Ejercicio 5
# Creo una lista de números y otra para guardar las posiciones encontradas
L = [3, 5, 13, 12, 1, 9]
resultado = []

# Recorro los índices de la lista
for i in range(len(L)):
    if L[i] % 3 == 0:
        resultado.append(i)  # Guardo el índice si el número es múltiplo de 3

# Muestro las posiciones encontradas
print(resultado)
````

````python
# Ejercicio 6
# Creo un diccionario con diferentes valores numéricos
d = {'a': 4.3, 'b': 1, 'c': 7.8, 'd': -5}

# Inicializo el valor máximo con infinito negativo
maximo = float("-inf")

# Verifico si el diccionario está vacío
if len(d) == 0:
    print("El diccionario está vacío")

# Recorro cada clave y valor para encontrar el mayor
for clave, valor in d.items():
    if valor > maximo:
        maximo = valor
        resultado = clave

# Muestro la clave que contiene el valor máximo
print(resultado)
````

````python
# Ejercicio 7
# Creo dos listas con diferentes valores
a = [2, 4, 6, 8]
b = [7, 11, 15, 22]

# Recorro los valores mayores que 5 de la primera lista
for i in a:
    if i > 5:
        # Multiplico por los valores menores que 14 de la segunda lista
        for j in b:
            if j < 14:
                print(f"{i} x {j} = {i * j}")
````

````python
# Ejercicio 8
# Solicito un valor e intento convertirlo a número
try:
    x = input()
    numero = float(x)

    # Realizo la división y muestro el resultado
    print(f"El resultado de 10/{x} es: {10 / numero}")

# Capturo el error si el valor ingresado no es numérico
except ValueError:
    print(f"El valor '{x}' no es numérico")

# Capturo el error si se intenta dividir entre cero
except ZeroDivisionError:
    print("Error: división por cero")
````

````python
# Ejercicio 9
# Creo un diccionario con las asignaturas y sus notas
notas = {
    'Programación': 9.5,
    'Inteligencia artificial': 7.6,
    'Redes': 5.2,
    'Sistemas operativos': 6.4
}

# Solicito una asignatura y busco su nota
try:
    asignatura = input("Introduce una asignatura: ")
    print(f"La nota de la asignatura '{asignatura}' es: {notas[asignatura]}")

# Capturo el error si la asignatura no existe
except KeyError:
    print(f"Asignatura incorrecta: {asignatura}")
````

````python
# Ejercicio 10
# Creo una lista con números enteros y decimales
lista = [1, 4, -3, -1.5, 6.5, 2, 8, 2.1]

# Selecciono únicamente los números enteros positivos
[numero for numero in lista if type(numero) == int and numero > 0]
````

````python
# Ejercicio 11
# Creo una palabra para buscar sus vocales
palabra = "murcielago"

# Creo un conjunto con las vocales encontradas, sin repetirlas
{letra for letra in palabra if letra in ('a', 'e', 'i', 'o', 'u')}
````

````python
# Ejercicio 12
# Creo una lista con los números del 0 al 50 que contienen el dígito 3
[numero for numero in range(51) if '3' in str(numero)]
````

````python
# Ejercicio 13
# Creo una frase para obtener la longitud de cada palabra
frase = "Soy un ser humano"

# Creo un diccionario con cada palabra y su cantidad de letras
{palabra: len(palabra) for palabra in frase.split()}
````

````python
# Ejercicio 14
# Creo un diccionario para convertir los números del 6 al 10 en palabras
numero_a_palabra = {6: "seis", 7: "siete", 8: "ocho", 9: "nueve", 10: "diez"}

# Mantengo los números del 1 al 5 y convierto del 6 al 10 en palabras
[numero if numero <= 5 else numero_a_palabra[numero]
 for numero in range(1, 11)]
````
