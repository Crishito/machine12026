# Series

* Datos unidimensionales (similar a NumPy)
* Elementos + índices modificables

```python
# Creación de una Serie de Pandas
countries = pd.Series(['Spain', 'Andorra', 'Gibraltar', 'Portugal', 'France'])
print(countries)
```

```python
# Creación de una Serie especificando un índice personalizado
# range(10, 60, 10) genera: [10, 20, 30, 40, 50]
countries = pd.Series(['Spain', 'Andorra', 'Gibraltar', 'Portugal', 'France'], 
                      index=range(10, 60, 10))
print(countries)
```

```python
# Los índices pueden ser de más tipos (etiquetas de texto)
football_cities = pd.Series(['Barcelona', 'Madrid', 'Valencia', 'Sevilla'],
                            index=['a', 'b', 'c', 'd'])
print(football_cities)
```

```python
# Atributos
football_cities.name = 'Ciudades con dos equipos en primera' # nombrar la Serie
football_cities.index.name = 'Id' # Describir los índices
print(football_cities)
```

```python
# Acceso a elementos en la Serie
# acceso por posición usando iloc (forma recomendada)
print(football_cities.iloc[2])

# acceso a través del índice semántico (etiqueta)
print(football_cities['c'])

# Verificación de igualdad entre ambos métodos de acceso
print(football_cities['c'] == football_cities.iloc[2])
```

# Tratamiento similar a ndarra

```python
# múltiple recolección de elementos
# acceso múltiple usando etiquetas del índice
print(football_cities.loc[['a', 'c']])

# acceso múltiple usando posiciones
print(football_cities.iloc[[0, 3]])
```

```python
# slicing
print(football_cities[:'c']) # incluye ambos extremos con el indice semantico
print(football_cities[:2])
```

```python
#cast a list
lista = list(football_cities[:'c'])
print(lista)
print(type(lista))
```

```python
# El resultado es un objeto tipo Series (subconjunto)
# Pandas incluye el límite final ('c') cuando usamos etiquetas
print(type(football_cities[:'c']))
```

```python
# cast a ndarray
import numpy as np

# Convertir una Serie de Pandas a un arreglo puro de NumPy (ndarray)
cities = np.array(football_cities[:'c']) 

print(cities)
print(type(cities)) # Verificamos que el tipo ahora es <class 'numpy.ndarray'>
```

```python
# cast a dictionary
# Convertir una Serie de Pandas a un diccionario de Python
# Las etiquetas (índices) se convierten en las claves y los valores se mantienen

lista = dict(football_cities[:'c'])

print(lista)
print(type(lista)) # Verificamos que ahora es <class 'dict'>
```

```python
# uso de masks para seleccionar
fibonacci = pd.Series([0, 1, 1, 2, 3, 5, 8, 13, 21])
print(fibonacci)

# Crear una máscara para seleccionar valores mayores a 10
mask = fibonacci > 10
print(mask)

# Seleccionar los valores que cumplen la condición de la máscara
print(fibonacci[mask])

# Crear una Serie con los valores esperados
dst = pd.Series([13,21])
print(dst)

# Comparar si ambas Series son iguales
dst.equals(fibonacci)

# Guardar en una nueva Serie los valores filtrados con la máscara
fb = fibonacci[mask]

# Reiniciar los índices de la Serie filtrada
fb.reset_index(drop=True, inplace=True)

print(fb)

# Comprobar si la Serie filtrada coincide con la Serie esperada
dst.equals(fb)
```

```python
# aplicar funciones de numpy a la serie

import numpy as np

# Sumar todos los valores de la Serie usando la función sum de NumPy
print(np.sum(fibonacci))
```

```python
# filtrado con np.where

distances = pd.Series([12.1, np.nan, 12.8, 76.9, 6.1, 7.2])

# Reemplazar los valores nulos por 2 y mantener los valores que no son nulos
valid_distances = np.where(pd.notnull(distances), distances, 2)

# Mostrar el resultado del filtrado
print(valid_distances)

# Verificar el tipo de dato del resultado
print(type(valid_distances))
```

### Iteración

```python
# iterar sobre elementos
for value in fibonacci:
    # Mostrar cada valor de la Serie
    print('Value: ' + str(value))

# iterar sobre indices
for index in fibonacci.index:
    # Mostrar cada índice de la Serie
    print('Index: ' + str(index))
```

```python
# iterar sobre elementos e índices al mismo tiempo

for index, value in fibonacci.items():
    # Mostrar el índice y el valor correspondiente de la Serie
    print('Index: ' + str(index) + ' Value: ' + str(value))
```

```python
# iterar sobre elementos e índices usando zip

for index, value in zip(fibonacci.index, fibonacci):
    # Mostrar el índice y el valor correspondiente de la Serie
    print('Index: ' + str(index) + ' Value: ' + str(value))
```

## Series como diccionarios

* Interpretar el índice como clave
* Acepta operaciones para diccionarios

```python
# crear una serie a partir de un diccionario

serie = pd.Series({'Carlos': 100, 'Marcos': 98})

# Mostrar los índices de la Serie
print(serie.index)

# Mostrar los valores de la Serie
print(serie.values)

# Mostrar la Serie completa
print(serie)

# Verificar el tipo de dato de la Serie
print(type(serie))
```

```python
# añade y elimina elementos a través de índices
# Añadir un nuevo elemento a la Serie usando un índice
serie['Pedro'] = 12

# Modificar el valor de un elemento existente
serie['Pedro'] = 15

# Eliminar un elemento de la Serie mediante su índice
del serie['Marcos']

# Mostrar la Serie actualizada
print(serie)
```

```python
# query una serie
# Consultar un valor de la Serie mediante su índice
# print(serie['Marcos'])

# Verificar si el índice 'Marcos' existe dentro de la Serie
if 'Marcos' in serie:
    # Mostrar el valor asociado al índice 'Marcos'
    print(serie['Marcos'])

# Mostrar la Serie completa
print(serie)
```

```python
# suma de dos series

# suma de valores con el mismo índice (NaN si no aparece en ambas)

serie1 = pd.Series([10,20,30,40], index=range(4))
serie2 = pd.Series([1,2,3], index=range(3))

# Sumar las dos Series, los valores con índices iguales se suman
# y los índices que no existen en ambas generan NaN
suma = serie1 + serie2

# Mostrar el resultado de la suma
print(suma)
```

```python
# resta de series (similar a la suma)

# Restar los valores de las Series con el mismo índice
# Si un índice no existe en ambas Series, el resultado será NaN
print(serie1 - serie2)
```

```python
# operaciones de pre-filtrado

result = serie1 + serie2

# Reemplazar los valores nulos (NaN) por 0 usando una máscara con isnull()
result[pd.isnull(result)] = 0

# Mostrar la Serie después del reemplazo
print(result)
```

### Diferencias entre Pandas Series y diccionario

* Diccionario, es una estructura que relaciona las claves y los valores de forma arbitraria.
* Series, estructura de forma estricta listas de valores con listas de índice asignado en la posición.
* Series, es más eficiente para ciertas operaciones que los dicionarios.
* En las Series los valores de entrada pueden ser listas o Numpy arrays.
* En Series los índices semánticos pueden ser integers o caracteres, en los valores igual.
* Series se podría entender entre una lista y un diccionario Python, pero es de una dimensión.

## Modificar DataFrames

```python
# añadir columnas

diccionario = {
    "Nombre": ["Marisa", "Laura", "Manuel"],
    "Edad": [34, 29, 12]
}

# Crear un DataFrame con columnas adicionales aunque no existan en el diccionario
frame = pd.DataFrame(diccionario, columns=['Nacionalidad', 'Nombre', 'Edad', 'Profesion', 'Direccion'])

# Añadir valores a la columna "Direccion" para todas las filas
frame['Direccion'] = 'Desconocida'

# Mostrar el DataFrame actualizado
display(frame)
```

```python
# crear una lista con direcciones

lista_direcciones = ['Rue 13 del Percebe, 13', 'Evergreen Terrace, 3', 'Av de los Rombos, 12']
```

```python
# añadir una lista de valores a la columna "Direccion"

# Asignar las direcciones de la lista a la columna del DataFrame
frame['Direccion'] = lista_direcciones

# Mostrar el DataFrame actualizado
display(frame)
```

```python
# añadir fila (requiere todos los valores)

# Crear una lista con los valores de la nueva fila
user_2 = ['Alemania', 'Klaus', 20, 'none', 'Desconocida']

# Añadir una nueva fila al DataFrame usando el índice 3
# La lista debe contener un valor para cada columna
frame.loc[3] = user_2

# Mostrar el DataFrame actualizado
display(frame)
```

```python
# eliminar fila (similar a Series)

frame = pd.DataFrame(diccionario, columns=['Nacionalidad', 'Nombre', 'Edad', 'Profesion'])

# Eliminar la fila con índice 2
# Necesitamos reasignar el frame porque drop() devuelve un nuevo DataFrame
# y no modifica el original a menos que usemos inplace=True
frame = frame.drop(2)

# Mostrar el DataFrame después de eliminar la fila
display(frame)

# Eliminar la columna "Nombre" usando axis=1 (indica que es una columna)
# inplace=True modifica directamente el DataFrame original
frame.drop('Nombre', axis=1, inplace=True)

# Mostrar el DataFrame actualizado
display(frame)
```

```python
# eliminar columna

# Eliminar la columna "Profesion" del DataFrame usando del
del frame['Profesion']

# Mostrar el DataFrame actualizado
display(frame)
```

```python
# acceder a la traspuesta (como una matriz)
display(frame.T)
```

```python
# iteración sobre el DataFrame?

frame = pd.DataFrame(diccionario, columns=['Nacionalidad', 'Nombre', 'Edad', 'Profesion'])

# Mostrar el DataFrame creado
display(frame)

# Iterar sobre un DataFrame recorre los nombres de las columnas
for a in frame:
    # 'a' representa el nombre de cada columna del DataFrame
    print(a)

    # Mostrar el tipo de dato de 'a' (es un string)
    print(type(a))
```

```python
# iteracion sobre filas

# Recorrer los valores de cada fila del DataFrame
for value in frame.values:
    # Mostrar los datos de la fila actual
    print(value)

    # Mostrar el tipo de dato de la fila (array de NumPy)
    print(type(value))
```

```python
# iterar sobre filas y luego sobre cada valor

# Recorrer cada fila del DataFrame
for values in frame.values:
    # Recorrer cada valor dentro de la fila actual
    for value in values:
        # Mostrar cada elemento de la fila
        print(value)
```

## Indexación y slicing con DataFrames

```python
# crear una lista de diccionarios para generar un DataFrame

d1 = {'ciudad': 'Valencia', 'temperatura': 10, 'o2': 1}
d2 = {'ciudad': 'Barcelona', 'temperatura': 8}
d3 = {'ciudad': 'Valencia', 'temperatura': 9}
d4 = {'ciudad': 'Madrid', 'temperatura': 10, 'humedad': 80}
d5 = {'ciudad': 'Sevilla', 'temperatura': 15, 'humedad': 50, 'co2': 6}
d6 = {'ciudad': 'Valencia', 'temperatura': 10, 'humedad': 90, 'co2': 10}

# Crear una lista con los diccionarios
ls_data = [d1, d2, d3, d4, d5, d6]

# Crear un DataFrame usando los diccionarios como filas
# Los índices de las filas serán las letras a, b, c, d, e, f
df_data = pd.DataFrame(ls_data, index=list('abcdef'))

# Mostrar el DataFrame creado
display(df_data)
```

```python
# Acceso a un valor concreto por índice posicional [fila, columna]
# iloc usa posiciones numéricas empezando desde 0
print(df_data.iloc[1,1])

# Acceso a todos los valores hasta un índice por enteros
# Selecciona las primeras 3 filas y las primeras 4 columnas
display(df_data.iloc[:3, :4])

# Acceso a datos de manera explícita usando índices semánticos (nombres)
# loc permite acceder usando etiquetas de filas y columnas

# Obtener el valor de la fila 'a' y columna 'temperatura'
display(df_data.loc['a', 'temperatura'])

# Obtener desde la fila 'a' hasta 'c' y columnas desde el inicio hasta 'o2'
# En loc los límites sí se incluyen
display(df_data.loc[:'c', :'o2'])

# Obtener desde la fila 'a' hasta 'c' y columnas entre 'temperatura' y 'o2'
display(df_data.loc[:'c', 'temperatura':'o2'])

# Obtener todas las filas y únicamente las columnas seleccionadas
display(df_data.loc[:, ['ciudad', 'o2']])
```

```python
# indexación con nombre de columna (por columnas)
# Acceder a una sola columna del DataFrame, devuelve una Serie
print(df_data['ciudad'])

# Acceder a varias columnas del DataFrame, devuelve un DataFrame
display(df_data[['ciudad', 'o2']])
```

```python
# Indexación con índice posicional (NO permitido para DataFrame)
# Al usar un número entre corchetes, Pandas lo interpreta como nombre de columna
# y no como posición de fila.
# Si no existe una columna llamada 0, generará un KeyError.
df_data[0]
```

```python
# indexar por posición con 'iloc'
# Acceder a la primera fila del DataFrame usando su posición
# Devuelve una Serie donde los índices son los nombres de las columnas
print(df_data.iloc[0])
```

```python
# indexar semántico con 'loc'
# Acceder a una fila usando su índice o etiqueta
# Devuelve una Serie donde los índices son los nombres de las columnas
df_data.loc['a']
```

```python
# indexar semántico con 'loc'
# Acceder a las filas desde el índice inicial hasta el índice 'b'
# loc incluye el límite final, por eso devuelve las filas con índices 'a' y 'b'
# El resultado es un DataFrame
df_data.loc[:'b']
```

```python
# slicing anidado
# Primero selecciona las filas desde el inicio hasta el índice 'b'
# Luego selecciona únicamente las columnas "o2" y "humedad"
df_data.loc[:'b'].loc[:, ["o2", "humedad"]]
```

```python
# si se modifica una porción del DataFrame se modifica el DataFrame original
# (referencia)

# Mostrar el DataFrame original
display(df_data)

# Obtener una fila del DataFrame como una Serie
serie = df_data.loc['a']
print(serie)

# Modificar un valor de la Serie obtenida
# Puede modificar el DataFrame original porque trabaja con una referencia
# Puede generar un warning de copia (SettingWithCopyWarning)
serie.iloc[2] = 3000

# Mostrar el DataFrame después de la modificación
display(df_data)

# copiar DataFrame

# Crear una copia independiente de la fila seleccionada
df_2 = df_data.loc['a'].copy()

# Modificar la copia sin afectar el DataFrame original
df_2.iloc[2] = 3000

# Mostrar la copia modificada
display(df_2)

# Mostrar el DataFrame original para comprobar que no cambió
display(df_data)
```

```python
# ambos aceptan 'axis' como argumento

# iloc con axis=1 permite acceder a columnas mediante su posición
# df_data.iloc(axis=1)[0] --> todos los valores asignados a la primera columna "ciudad"

# loc con axis=1 permite acceder a columnas mediante su nombre
# Equivale a usar df_data['ciudad']
df_data.loc(axis=1)['ciudad']
```

```python
# qué problema puede tener este fragmento?

frame = pd.DataFrame(
    {"Name": ['Carlos', 'Pedro'], "Age": [34, 22]},
    index=[1, 0]
)

# El problema es que los índices están desordenados.
# Esto puede causar confusión al acceder a los datos porque las filas
# no siguen el orden habitual de creación.
display(frame)
```

```python
# por defecto, pandas interpreta índice posicional --> error en frames
# cuando hay posible ambigüedad, utilizar loc y iloc

# Acceder a la primera fila por posición (posición 0)
# iloc trabaja con índices numéricos de posición
print('Primera fila\n')
print(frame.iloc[0])

# Acceder a la fila cuyo índice es 0
# loc trabaja con las etiquetas o índices definidos en el DataFrame
print('\nElemento con index 0\n')
print(frame.loc[0])
```

## Objeto Index de Pandas

```python
# Construcción de índices

# Crear un objeto Index de Pandas con una lista de valores
ind = pd.Index([2, 3, 5, 23, 26])

# Recuperar un valor del índice mediante su posición
print(ind[3])

# Recuperar valores del índice usando slicing
print(ind[::2])
```

```python
# change index column

# Crear un DataFrame usando el objeto Index como índice inicial
frame = pd.DataFrame(
    {
        "Name": ['Carlos', 'Pedro', 'Manolo', 'Luis', 'Alberto'],
        "Age": [34, 22, 15, 55, 23]
    },
    index=ind
)

# Mostrar el DataFrame inicial
display(frame)

# Cambiar la columna "Age" para que sea el nuevo índice del DataFrame
# inplace=True aplica el cambio directamente sobre el DataFrame
frame.set_index('Age', inplace=True)

# Mostrar el DataFrame con el nuevo índice
display(frame)
```

## Slicing

```python
# slice por filas

# Crear un diccionario con información de personajes
d_and_d_characters = {
    'Name': ['bundenth', 'theorin', 'barlok'],
    'Strength': [10, 12, 19],
    'Wisdom': [20, 13, 6]
}

# Crear un DataFrame con índices personalizados
character_data = pd.DataFrame(d_and_d_characters, index=['a', 'b', 'c'])

# Mostrar el DataFrame creado
display(character_data)

# Seleccionar todas las filas excepto la última
display(character_data[:-1])

# Seleccionar la fila con posición 1
# El límite final no se incluye en el slicing
display(character_data[1:2])
```

```python
# slicing para columnas
# Seleccionar únicamente las columnas "Name" y "Wisdom" del DataFrame
display(character_data[['Name', 'Wisdom']])
```

```python
# slicing con 'loc' e 'iloc'
# Seleccionar filas desde la posición 1 hasta el final usando iloc
# iloc trabaja con posiciones numéricas
display(character_data.iloc[1:])

# Seleccionar filas desde el inicio hasta el índice 'b'
# y columnas desde 'Name' hasta 'Strength' usando loc
# loc trabaja con etiquetas e incluye el límite final
display(character_data.loc[:'b', 'Name':'Strength'])
```

¿Cómo filtrar filas y columnas? Por ejemplo, para todos los personajes, obtener ‘Name’ y ‘Strength’

```python
# usando 'loc' para hacer slicing
display(character_data.loc[:,'Name':'Strength'])
```

```python
# usando 'loc' para buscar específicamente filas y columnas
display(character_data.loc[ ['a','c'], ['Name','Wisdom'] ])
```

```python
# lo mismo con 'iloc'?

# Seleccionar filas y columnas específicas mediante sus posiciones
# iloc trabaja con índices numéricos de posición

# Seleccionar filas 0 y 2, y columnas 0 y 2
display(character_data.iloc[[0, 2], [0, 2]])

# Seleccionar la primera y última fila, y la primera y última columna
display(character_data.iloc[[0, -1], [0, -1]])
```

```python
# lista de los personajes con el atributo Strength > 11
display(character_data.loc[character_data['Strength'] > 11, ['Name', 'Strength']])
```

```python
# listar los personajes con Strength > 15 o Wisdom > 15
display(character_data.loc[(character_data['Strength'] > 15) | (character_data['Wisdom'] > 15)])
```

```python
# importar librerías necesarias

import numpy as np
import pandas as pd
import zipfile  # para descomprimir archivos zip
import urllib.request  # para descargar archivos desde una URL
import os

# descargar el dataset MovieLens

# URL del archivo comprimido del dataset
url = 'http://files.grouplens.org/datasets/movielens/ml-1m.zip'

# Crear la ruta donde se guardará el archivo descargado
ruta = os.path.join("res", "ml-1m.zip")

# Descargar el archivo desde la URL y guardarlo en la ruta indicada
urllib.request.urlretrieve(url, ruta)
```

```python
# descomprimiendo archivo zip
# Definir la ruta donde se extraerán los archivos
ruta_ext = os.path.join("res")

# Abrir el archivo ZIP en modo lectura
with zipfile.ZipFile(ruta, 'r') as z:
    # Mostrar mensaje de inicio de extracción
    print('Extracting all files...')

    # Extraer todos los archivos en la ruta indicada
    z.extractall(ruta_ext)

    # Mostrar mensaje de finalización
    print('Done!')

# take a look at readme y revisar formatos
# Revisar el archivo readme para conocer la estructura y formatos de los datos
```

```python
# filtrar edad incorrecta (mínimo 18) Fixing it by Copying the slice
users_dataset = pd.read_csv(ruta_users, sep='::', index_col=0,
header=None, names=['UserID','Gender','Age','Occupation','Zip-code'],
engine='python')

under_age = users_dataset[users_dataset['Age'] == 1]

# crear una copia de los datos filtrados
under_age_copy = under_age.copy()

display(under_age_copy.head())

# reemplazar la edad incorrecta por valores nulos
under_age_copy['Age'] = np.nan

display(under_age_copy.head())

# actualizar los registros en la tabla original
users_dataset[users_dataset['Age'] == 1] = under_age_copy

display(users_dataset.head())
```

```python
# filtrar edad incorrecta (mínimo 18) Remove them from the dataset
users_dataset = pd.read_csv(ruta_users, sep='::', index_col=0,
header=None, names=['UserID','Gender','Age','Occupation','Zip-code'],
engine='python')

display(users_dataset[users_dataset['Age'] == 1].head(4))

# reemplazar edades incorrectas por valores nulos
users_dataset.loc[users_dataset['Age'] == 1, 'Age'] = np.nan

display(users_dataset)
display(users_dataset.loc[pd.isnull(users_dataset['Age'])].head(4))

# eliminar registros con valores nulos en la edad
users_dataset.drop(users_dataset[pd.isnull(users_dataset['Age'])].index,
inplace=True)

display(users_dataset.head(4))
```

* Hacer un análisis general de los otros dos archivos CSV en ml-1m (‘movies.dat’ y ‘ratings.dat’)
* Analizando el dataset ratings.dat, ¿hay algún usuario que no tenga ninguna review? ¿Cuántos tienen menos de 30 reviews?

```python
# especificar nombres, cargar sin cabecera
import os
import pandas as pd

ruta_ratings = os.path.join("res", "ml-1m", "ratings.dat")

# cargar datos de calificaciones
ratings_dataset = pd.read_csv(ruta_ratings, sep='::',
header=None, names=['UserID','MovieID','Rating','Timestamp'],
engine='python')

display(ratings_dataset)
```

```python
# hay algún usuario que no tenga ninguna review
n_users_with_reviews = len(ratings_dataset.groupby(by='UserID'))
print(n_users_with_reviews)

n_users = len(users_dataset)
print(n_users)

print(n_users - n_users_with_reviews)
```

```python
# Cuántos tienen menos de 30 reviews
ratings_under_30 = ratings_dataset.groupby(['UserID'])['UserID'].filter(lambda x: len(x) < 30).value_counts()

display(ratings_under_30)
print(len(ratings_under_30))
```
