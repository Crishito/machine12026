# Operaciones en pandas

**Búsqueda**

```python
# crear una matriz aleatoria y convertirla en un DataFrame
rand_matrix = np.random.randint(6, size=(2,3))
frame = pd.DataFrame(rand_matrix, columns=list('ABC'))

display(frame)
```

```python
# buscando columnas (DataFrame como dic, busca en claves)
'A' in frame
```

```python
# buscando valores
display(frame.isin([3,2])) # --> mask de respuesta (valores que son 3 o 2)
```

```python
# Contar el número de ocurrencias
print(frame.isin([4]).values.sum())

# devuelve una máscara con valores True si el elemento es 4
display(frame.isin([4]))

print(frame.isin([4]).values)

# pandas es una capa alrededor de numpy
type(frame.isin([4]).values)
```

```python
# Cuántos valores son >= 2
mask = frame >= 2
print(mask.values.sum())

display(mask)
```

**Ordenación**

```python
# crear una matriz aleatoria con índices mezclados
from random import shuffle

rand_matrix = np.random.randint(20, size=(5,4))

indices = list(range(5))
shuffle(indices) # mezcla indices

frame = pd.DataFrame(rand_matrix, columns=list('DACB'), index=indices)

display(frame)
```

```python
# ordenar por índice
display(frame.sort_index(ascending=False))
display(frame)
```

```python
#ordenar por columna
display(frame.sort_index(axis=1, ascending=False))
```

```python
# ordenar filas por valor en columna
display(frame.sort_values(by='A', ascending=True))
```

```python
# ordenar columnas por valor en fila
display(frame.sort_values(by=1, axis=1, ascending=True))
```

```python
# ordenar por valor en columna y guardar cambios
frame.sort_values(by='A', ascending=False, inplace=True)
display(frame)
```

**Ranking**
  
* Construir un ranking de valores

```python
# mostrar el DataFrame
display(frame)
```

```python
# asignar rangos a los valores del DataFrame por fila
display(frame.rank(method='max', axis=1))
```

```python
# Imprimir los valores de la columna 'C' ordenados de mayor a menor
for x in frame.sort_values(by='C', ascending=False)['C'].values:
    print(x)
```

# Operaciones

Operaciones matemáticas entre objetos

```python
# crear dos matrices aleatorias y convertirlas en DataFrames
matrixA = np.random.randint(100, size=(4,4))
matrixB = np.random.randint(100, size=(4,4))

frameA = pd.DataFrame(matrixA)
frameB = pd.DataFrame(matrixB)

display(frameA)
display(frameB)
```

```python
# sumar DataFrames usando operadores y métodos
display(frameA + frameB == frameA.add(frameB))

display(frameA + frameB)
```

```python
# restar DataFrames usando operadores y métodos
display(frameB - frameA == frameB.sub(frameA))

display(frameB - frameA)
```

```python
# si los DataFrames no son iguales, valor por defecto NaN
frameC = pd.DataFrame(np.random.randint(100, size=(3,3)))

display(frameA)
display(frameC)
display(frameC + frameA)
```

```python
# especificar el valor por defecto con el argumento fill_value
display(frameA.add(frameC, fill_value=0))
```

Operadores aritméticos solo válidos en elementos aceptables

```python
# crear un DataFrame con valores de texto
frameD = pd.DataFrame({0: ['a','b'], 1: ['d','f']})

display(frameD)

# operación entre DataFrames con tipos incompatibles
# error al operar con DataFrames de tipos diferentes (números y texto)
frameA - frameD
```

Operaciones entre Series y DataFrames

```python
# crear un DataFrame con valores aleatorios
rand_matrix = np.random.randint(10, size=(3, 4))

df = pd.DataFrame(rand_matrix, columns=list('ABCD'))

display(df)

display(df.iloc[0])
display(type(df.iloc[0]))

# restar una fila al resto del DataFrame
display(df - df.iloc[0])
display(df.sub(df.iloc[0], axis=1))

# restar por columnas
display(df.sub(df['A'], axis=0))
```

Aplicación de funciones a medida con lambda

```python
# Generación de una matriz aleatoria y creación de un DataFrame con nombres de columnas.
rand_matrix = np.random.randint(10, size=(3, 4))

frame = pd.DataFrame(rand_matrix , columns=list('ABCD'))
display(frame)

# Aplicación de una función para obtener la diferencia entre el valor máximo y mínimo por fila.
print(frame.apply(lambda x: x.max() - x.min(), axis=1))
```

```python
# Definición de una función para calcular la diferencia entre el valor máximo y mínimo.
def max_min(x):
    return x.max() - x.min()

# Aplicación de la función para obtener la diferencia por columna.
print(frame.apply(max_min, axis=0))
```

```python
# Generación de una matriz aleatoria y creación de un DataFrame para trabajar con sus datos.
rand_matrix = np.random.randint(10, size=(3, 4))

frame = pd.DataFrame(rand_matrix, columns=list('ABCD'))

display(frame)

# Cálculo de la diferencia entre el valor máximo y mínimo de cada fila.
print(frame.apply(lambda x: x.max() - x.min(), axis=1)) # diferencia por fila
```

```python
# Creación de un DataFrame a partir de un diccionario con información de personas.
diccionario = { 
    "nombre": ["Marisa", "Laura", "Manuel", "Carlos"], 
    "edad": [34, 34, 11, 30],
    "puntos": [98, 12, 98, np.nan], 
    "genero": ["F", "F", "M", "M"] 
}

frame = pd.DataFrame(diccionario)

display(frame)

# Visualización de estadísticas generales de los datos numéricos del DataFrame.
display(frame.describe()) # datos generales de elementos
```

```python
# Operadores básicos para realizar cálculos sobre los datos del DataFrame.
print(frame.sum())

display(frame)

# Suma de los valores numéricos por cada fila del DataFrame.
print(frame.sum(axis=1, numeric_only=True))
```

```python
# Cálculo del promedio de los valores numéricos del DataFrame.
frame.mean(numeric_only=True)
```

```python
# Cálculo acumulativo de los valores del DataFrame.
frame.cumsum()
```

```python
# Conteo de valores disponibles en cada fila del DataFrame.
frame.count(axis=1)
```

```python
# Cálculo de la desviación estándar de una columna del DataFrame.
print(frame['edad'].std())
```

```python
# Obtención del índice correspondiente al valor máximo de una columna del DataFrame.
frame['edad'].idxmax()
```

```python
# Obtención del índice correspondiente al valor mínimo de una columna del DataFrame.
frame['puntos'].idxmin()
```

```python
# Filtrado del DataFrame para obtener las filas que contienen los valores máximos de una columna.
print(frame['puntos'].max())

display(frame[frame['puntos'] == frame['puntos'].max()])
```

```python
# Creación de una nueva columna con la posición o ranking de los valores de una columna.
frame["ranking"] = frame["puntos"].rank(method='max')
```

```python
# Visualización del DataFrame actualizado.
display(frame)
```

## Agregaciones

```python
# Visualización del DataFrame y agrupación de datos según una categoría.
display(frame)

df = frame.groupby('genero').count()

display(df)
```

```python
# Agrupación de datos descartando valores nulos.
df = frame.groupby('puntos').count()

display(df)
```

```python
# Cálculo del promedio de los valores numéricos agrupados por una categoría.
display(frame.groupby('genero').mean(numeric_only=True))
```

```python
# Obtención de los valores máximos de los datos agrupados por una categoría.
display(frame.groupby('genero').max())
```

```python
# Aplicación de funciones de agregación para obtener diferentes estadísticas agrupadas.
display(frame.groupby('genero')[['edad', 'puntos']].aggregate(['min', 'mean', 'max']))
```

```python
# Filtrado de los datos agrupados según una condición determinada.
def media(x):
    return x["edad"].mean() > 30

display(frame)

frame.groupby('genero').filter(media)
```

## Correlaciones

pandas incluye métodos para analizar correlaciones - Relación matemática entre dos variables (-1 negativamente relacionadas, 1 positivamente relacionadas, 0 sin relación) - obj.corr(obj2) –> medida de correlación entre los datos de ambos objetos - https://blogs.oracle.com/ai-anddatascience/post/introduction-to-correlation

### Ejemplo Fuel efficiency

* https://archive.ics.uci.edu/ml/datasets/Auto+MPG

```python
# Importación de Pandas y carga de un conjunto de datos externo en un DataFrame.
import pandas as pd

path = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'

mpg_data = pd.read_csv(path, sep='\s+', header=None,
                       names=['mpg', 'cilindros', 'desplazamiento', 'potencia',
                              'peso', 'aceleracion', 'año', 'origen', 'nombre'],
                       na_values='?', engine='c')
```

```python
# Visualización de una muestra aleatoria de registros del DataFrame.
display(mpg_data.sample(5))
```

```python
# Visualización de estadísticas generales incluyendo datos numéricos y categóricos.
display(mpg_data.describe(include='all'))
```

```python
# Visualización de la información general y estructura del DataFrame.
mpg_data.info()
```

### Correlaciones entre valores

```python
# Cálculo de la correlación entre las variables numéricas del DataFrame.
mpg_data.corr(numeric_only=True)
```

```python
# Año y origen no parecen tener una relación significativa con las demás variables.
# Eliminación de columnas antes de calcular la matriz de correlación.
corr_data = mpg_data.drop(['año', 'origen'], axis=1).corr(numeric_only=True)

display(corr_data)
```

```python
# Importación de la librería para la creación de gráficos.
import matplotlib.pyplot as plt
```

```python
# Representación gráfica de la matriz de correlación mediante una escala de colores.
corr_data.style.background_gradient(cmap=plt.get_cmap('RdYlGn'), axis=1)
```

```python
# Obtención de las variables con la correlación más negativa.
mpg_data.drop(['año', 'origen'], axis=1).corr(numeric_only=True).idxmin()
```

```python
# Obtención de las variables con la correlación más positiva.
mpg_data.drop(['año', 'origen'], axis=1).corr(numeric_only=True).idxmax() # consigo misma....
```

```python
# tabla similar con las correlaciones más positivas (evitar parejas del mismo valor)
positive_corr = mpg_data.drop(['año','origen'],axis=1).corr(numeric_only=True).copy()
np.fill_diagonal(positive_corr.values, 0)
display(positive_corr)
positive_corr.idxmax()
```

```python
# Representación gráfica de la matriz de correlación mediante una escala de colores ajustada.
positive_corr.style.background_gradient(cmap=plt.get_cmap('RdYlGn'), axis=1, vmin=-1.0, vmax=1.0)
```
